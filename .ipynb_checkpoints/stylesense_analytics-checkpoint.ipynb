# StyleSense Sales Analytics
## Retail Performance Analysis — Jan 2024 to Jun 2025

This notebook answers eight core business questions using **descriptive**, **diagnostic**, **predictive**, and **prescriptive** analytics across StyleSense's sales, customer, product, inventory, and expense data.

---

| Business Question | Analytics Type | Datasets |
|---|---|---|
| What's our most profitable product? | Descriptive | Sales + Products |
| Which customer segment spends most? | Descriptive | Customers + Sales |
| Do discounts actually increase sales? | Diagnostic | Sales |
| When should we stock winter vs summer products? | Predictive | Products + Inventory |
| Online vs Physical — which channel performs better? | Descriptive + Diagnostic | Sales |
| Are we spending marketing money wisely? | Diagnostic + Prescriptive | Sales + Expenses |
| Which products are frequently out of stock? | Diagnostic + Prescriptive | Inventory |
| What is our true profit after all expenses? | Descriptive + Prescriptive | Sales + Expenses |

## Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams['figure.dpi'] = 110
plt.rcParams['font.size'] = 10

BLUE   = "#1f77b4"
GREEN  = "#2ca02c"
ORANGE = "#ff7f0e"
RED    = "#d62728"
PURPLE = "#9467bd"
```

```python
# Load cleaned datasets
sales     = pd.read_csv('sales_clean_data.csv',     parse_dates=['date'])
customers = pd.read_csv('customers_clean_data.csv', parse_dates=['first_purchase_date','last_purchase_date'])
products  = pd.read_csv('products_clean_data.csv')
expenses  = pd.read_csv('expenses_clean_data.csv',  parse_dates=['date'])
inventory = pd.read_csv('inventory_clean_data.csv', parse_dates=['date'])

print("Datasets loaded:")
print(f"  sales     : {sales.shape}")
print(f"  customers : {customers.shape}")
print(f"  products  : {products.shape}")
print(f"  expenses  : {expenses.shape}")
print(f"  inventory : {inventory.shape}")
```

---
## Q1 — What's Our Most Profitable Product?
**Analytics type: Descriptive**

We measure profitability at two levels:
- **Net profit per transaction** (`net_profit`) — actual realised profit after returns
- **Profit margin %** from the products catalogue — tells us which products earn more per dollar sold

```python
# Total net profit and units sold per product
product_profit = sales.groupby('product_name').agg(
    total_net_profit = ('net_profit', 'sum'),
    total_units_sold = ('quantity_sold', 'sum'),
    total_revenue    = ('total_sale', 'sum'),
    transactions     = ('sale_id', 'count')
).reset_index()

product_profit['avg_profit_per_unit'] = (product_profit['total_net_profit'] / product_profit['total_units_sold']).round(2)
product_profit = product_profit.sort_values('total_net_profit', ascending=False)

print("Top 10 Most Profitable Products (by total net profit):")
print(product_profit.head(10)[['product_name','total_net_profit','total_units_sold','avg_profit_per_unit']].to_string(index=False))
```

```python
# Bar chart: top 10 products by net profit
top10 = product_profit.head(10)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Left: total net profit
axes[0].barh(top10['product_name'][::-1], top10['total_net_profit'][::-1], color=BLUE)
axes[0].set_xlabel("Total Net Profit ($)")
axes[0].set_title("Top 10 Products by Total Net Profit")
axes[0].xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"${x:,.0f}"))

# Right: profit margin % from product catalogue
prod_margins = products[['product_name','profit_margin_pct']].sort_values('profit_margin_pct', ascending=False)
axes[1].barh(prod_margins['product_name'][::-1], prod_margins['profit_margin_pct'][::-1], color=GREEN)
axes[1].set_xlabel("Profit Margin (%)")
axes[1].set_title("Profit Margin % by Product (Catalogue)")
axes[1].xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"{x:.0f}%"))

plt.tight_layout()
plt.show()
```

```python
# Profit breakdown by category
category_profit = sales.groupby('category').agg(
    total_net_profit = ('net_profit', 'sum'),
    total_revenue    = ('total_sale', 'sum')
).reset_index()
category_profit['profit_margin_pct'] = (category_profit['total_net_profit'] / category_profit['total_revenue'] * 100).round(1)
category_profit = category_profit.sort_values('total_net_profit', ascending=False)

print("Profitability by Category:")
print(category_profit.to_string(index=False))
```

**Insight:** The products with the highest total net profit are not always those with the highest margin %.
High-margin products sold in low volume can be overtaken by mid-margin products with strong sales velocity.
The category breakdown reveals which clothing lines drive the most absolute profit for StyleSense.

## Q2 — Which Customer Segment Spends Most?
**Analytics type: Descriptive**

We look at both the **registration-based segment** (Loyal, New, Seasonal, Occasional) from the CRM
and the **behaviour-based derived_segment** we engineered during cleaning — to give a full picture.

```python
# Average and total spend by CRM segment
seg_spend = customers.groupby('segment').agg(
    customer_count   = ('customer_id', 'count'),
    avg_spent        = ('total_spent', 'mean'),
    total_spent      = ('total_spent', 'sum'),
    avg_orders       = ('num_orders', 'mean')
).reset_index()
seg_spend['avg_spent']  = seg_spend['avg_spent'].round(2)
seg_spend['avg_orders'] = seg_spend['avg_orders'].round(1)
seg_spend = seg_spend.sort_values('avg_spent', ascending=False)

print("Spending by CRM Segment:")
print(seg_spend.to_string(index=False))
```

```python
# Spending by behaviour-based derived_segment
derived_spend = customers.groupby('derived_segment').agg(
    customer_count = ('customer_id', 'count'),
    avg_spent      = ('total_spent', 'mean'),
    total_spent    = ('total_spent', 'sum'),
    avg_orders     = ('num_orders', 'mean')
).reset_index()
derived_spend['avg_spent']  = derived_spend['avg_spent'].round(2)
derived_spend['avg_orders'] = derived_spend['avg_orders'].round(1)
derived_spend = derived_spend.sort_values('avg_spent', ascending=False)

print("Spending by Behaviour-Based Segment:")
print(derived_spend.to_string(index=False))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Left: CRM segment avg spend
axes[0].bar(seg_spend['segment'], seg_spend['avg_spent'], color=PURPLE)
axes[0].set_title("Average Spend by CRM Segment")
axes[0].set_ylabel("Avg Spend ($)")
axes[0].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"${x:,.0f}"))
for bar, val in zip(axes[0].patches, seg_spend['avg_spent']):
    axes[0].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 20, f"${val:,.0f}", ha='center', fontsize=9)

# Right: Behaviour segment avg spend
axes[1].bar(derived_spend['derived_segment'], derived_spend['avg_spent'], color=ORANGE)
axes[1].set_title("Average Spend by Behaviour Segment")
axes[1].set_ylabel("Avg Spend ($)")
axes[1].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"${x:,.0f}"))
for bar, val in zip(axes[1].patches, derived_spend['avg_spent']):
    axes[1].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 20, f"${val:,.0f}", ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# Top 5 favourite categories per segment
seg_fav = customers.groupby(['segment','favourite_category']).size().reset_index(name='count')
seg_fav = seg_fav.sort_values(['segment','count'], ascending=[True, False])
top_fav = seg_fav.groupby('segment').head(3)
print("Top 3 Favourite Categories per Segment:")
print(top_fav.to_string(index=False))
```

**Insight:** Loyal customers drive the highest lifetime value. However the behaviour-based segment
shows that High Engagement customers — regardless of CRM label — are the most valuable cohort.
Marketing should prioritise converting Mid Engagement customers upward, and target Loyal/High Engagement
customers with premium product promotions.

## Q3 — Do Discounts Actually Increase Sales?
**Analytics type: Diagnostic**

This question asks whether applying a discount leads to more units sold and better revenue — or whether
it simply reduces margin without meaningfully lifting volume. We compare volume, revenue, and profit
across each discount tier (0%, 5%, 10%, 15%, 20%).

```python
# Aggregate metrics by discount tier
discount_analysis = sales.groupby('discount_pct').agg(
    transactions     = ('sale_id', 'count'),
    avg_units_sold   = ('quantity_sold', 'mean'),
    total_units_sold = ('quantity_sold', 'sum'),
    total_revenue    = ('total_sale', 'sum'),
    total_net_profit = ('net_profit', 'sum'),
    return_rate      = ('is_returned', 'mean')
).reset_index()

discount_analysis['avg_revenue_per_tx']  = (discount_analysis['total_revenue'] / discount_analysis['transactions']).round(2)
discount_analysis['avg_profit_per_tx']   = (discount_analysis['total_net_profit'] / discount_analysis['transactions']).round(2)
discount_analysis['profit_margin_pct']   = (discount_analysis['total_net_profit'] / discount_analysis['total_revenue'] * 100).round(1)
discount_analysis['return_rate']         = (discount_analysis['return_rate'] * 100).round(1)

print("Discount Impact Summary:")
print(discount_analysis[['discount_pct','transactions','avg_units_sold','avg_revenue_per_tx',
                          'avg_profit_per_tx','profit_margin_pct','return_rate']].to_string(index=False))
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 10))

x = discount_analysis['discount_pct'].astype(str) + '%'

# 1. Transaction volume
axes[0,0].bar(x, discount_analysis['transactions'], color=BLUE)
axes[0,0].set_title("Number of Transactions by Discount Tier")
axes[0,0].set_ylabel("Transactions")

# 2. Avg revenue per transaction
axes[0,1].bar(x, discount_analysis['avg_revenue_per_tx'], color=GREEN)
axes[0,1].set_title("Avg Revenue per Transaction by Discount Tier")
axes[0,1].set_ylabel("Avg Revenue ($)")
axes[0,1].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"${x:,.0f}"))

# 3. Profit margin %
axes[1,0].bar(x, discount_analysis['profit_margin_pct'], color=ORANGE)
axes[1,0].set_title("Profit Margin % by Discount Tier")
axes[1,0].set_ylabel("Margin (%)")
axes[1,0].axhline(discount_analysis['profit_margin_pct'].mean(), color=RED, linestyle='--', label='Average')
axes[1,0].legend()

# 4. Return rate
axes[1,1].bar(x, discount_analysis['return_rate'], color=RED)
axes[1,1].set_title("Return Rate % by Discount Tier")
axes[1,1].set_ylabel("Return Rate (%)")

plt.suptitle("Impact of Discounts on Sales Performance", fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

**Insight (Diagnostic):** If transaction volume rises at higher discounts but profit margin falls,
discounts are growing top-line revenue at the cost of profitability.
If return rates are higher on discounted items, that further erodes net profit.
The data here tells us whether StyleSense's discount strategy is value-creating or margin-destroying.

## Q4 — When Should We Stock Winter vs Summer Products?
**Analytics type: Predictive**

Using 18 months of inventory data, we identify when stock levels drop fastest for seasonal products.
This tells us the optimal restock window before demand peaks — giving actionable timing for buyers.

```python
# Tag each inventory record with month and season
inventory['month'] = inventory['date'].dt.month
inventory['month_name'] = inventory['date'].dt.strftime('%b')
inventory['year'] = inventory['date'].dt.year

# Merge with products to get season tag
inv_products = inventory.merge(products[['product_id','season']], on='product_id', how='left')

# Separate seasonal products
winter_inv = inv_products[inv_products['season'].isin(['Winter','All'])]
summer_inv = inv_products[inv_products['season'].isin(['Summer','All'])]

print("Winter products:", inv_products[inv_products['season'] == 'Winter']['product_name'].unique())
print("Summer products:", inv_products[inv_products['season'] == 'Summer']['product_name'].unique())
```

```python
# Average stock level by month for winter vs summer products
winter_monthly = winter_inv[winter_inv['season'] == 'Winter'].groupby('month')['stock_level'].mean().reset_index()
summer_monthly = summer_inv[summer_inv['season'] == 'Summer'].groupby('month')['stock_level'].mean().reset_index()

winter_monthly['month_name'] = pd.to_datetime(winter_monthly['month'], format='%m').dt.strftime('%b')
summer_monthly['month_name'] = pd.to_datetime(summer_monthly['month'], format='%m').dt.strftime('%b')

print("Average stock level by month — Winter products:")
print(winter_monthly[['month_name','stock_level']].to_string(index=False))
print()
print("Average stock level by month — Summer products:")
print(summer_monthly[['month_name','stock_level']].to_string(index=False))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

axes[0].plot(winter_monthly['month_name'], winter_monthly['stock_level'], marker='o', color=BLUE, linewidth=2)
axes[0].set_title("Avg Stock Level by Month — Winter Products")
axes[0].set_ylabel("Avg Stock Level (units)")
axes[0].set_xlabel("Month")
axes[0].axhline(40, color=RED, linestyle='--', label='Reorder Point (40)')
axes[0].legend()

axes[1].plot(summer_monthly['month_name'], summer_monthly['stock_level'], marker='o', color=ORANGE, linewidth=2)
axes[1].set_title("Avg Stock Level by Month — Summer Products")
axes[1].set_ylabel("Avg Stock Level (units)")
axes[1].set_xlabel("Month")
axes[1].axhline(40, color=RED, linestyle='--', label='Reorder Point (40)')
axes[1].legend()

plt.suptitle("Seasonal Stock Depletion Patterns", fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Near-stockout months for seasonal products
near_stockout_seasonal = inv_products[
    (inv_products['stock_level'] <= inv_products['reorder_point']) &
    (inv_products['season'].isin(['Winter','Summer']))
]

stockout_by_season_month = near_stockout_seasonal.groupby(
    ['season','month']
).size().reset_index(name='near_stockout_days')

stockout_by_season_month['month_name'] = pd.to_datetime(stockout_by_season_month['month'], format='%m').dt.strftime('%b')
print("Near-stockout events by season and month:")
print(stockout_by_season_month[['season','month_name','near_stockout_days']].to_string(index=False))
```

**Insight (Predictive):** Months where near-stockout events peak for winter products signal when
demand outpaces supply — these are the months StyleSense should complete restocking *before*.
The same logic applies to summer products. Buyers should place orders 4–6 weeks before the months
where stock levels historically drop to the reorder threshold.

## Q5 — Online vs Physical: Which Channel Performs Better?
**Analytics type: Descriptive + Diagnostic**

We compare the two channels across revenue, profit, order size, return rate, and discount usage
to understand which channel is more valuable and why.

```python
# Channel comparison
channel = sales.groupby('store_channel').agg(
    transactions     = ('sale_id', 'count'),
    total_revenue    = ('total_sale', 'sum'),
    total_net_profit = ('net_profit', 'sum'),
    avg_order_value  = ('total_sale', 'mean'),
    avg_discount_pct = ('discount_pct', 'mean'),
    return_rate      = ('is_returned', 'mean'),
    avg_units_sold   = ('quantity_sold', 'mean')
).reset_index()

channel['profit_margin_pct'] = (channel['total_net_profit'] / channel['total_revenue'] * 100).round(1)
channel['return_rate']       = (channel['return_rate'] * 100).round(1)
channel['avg_order_value']   = channel['avg_order_value'].round(2)
channel['avg_discount_pct']  = channel['avg_discount_pct'].round(1)

print("Channel Performance Summary:")
print(channel.to_string(index=False))
```

```python
# Monthly revenue trend by channel
sales['month'] = sales['date'].dt.to_period('M')
monthly_channel = sales.groupby(['month','store_channel'])['total_sale'].sum().reset_index()
monthly_channel['month_dt'] = monthly_channel['month'].dt.to_timestamp()

online   = monthly_channel[monthly_channel['store_channel'] == 'Online']
physical = monthly_channel[monthly_channel['store_channel'] == 'Physical']

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Left: monthly revenue trend
axes[0].plot(online['month_dt'], online['total_sale'], label='Online', color=BLUE, marker='o')
axes[0].plot(physical['month_dt'], physical['total_sale'], label='Physical', color=ORANGE, marker='o')
axes[0].set_title("Monthly Revenue: Online vs Physical")
axes[0].set_ylabel("Revenue ($)")
axes[0].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"${x:,.0f}"))
axes[0].legend()
axes[0].tick_params(axis='x', rotation=45)

# Right: KPI bar comparison
metrics = ['total_revenue','total_net_profit','avg_order_value']
labels  = ['Total Revenue','Total Net Profit','Avg Order Value']
x = range(len(metrics))
width = 0.35
online_vals   = [channel[channel['store_channel']=='Online'][m].values[0] for m in metrics]
physical_vals = [channel[channel['store_channel']=='Physical'][m].values[0] for m in metrics]

axes[1].bar([i - width/2 for i in x], online_vals,   width, label='Online',   color=BLUE)
axes[1].bar([i + width/2 for i in x], physical_vals, width, label='Physical', color=ORANGE)
axes[1].set_xticks(list(x))
axes[1].set_xticklabels(labels)
axes[1].set_title("Key Metrics: Online vs Physical")
axes[1].yaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f"${v:,.0f}"))
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
# Payment method split by channel
payment_channel = sales.groupby(['store_channel','payment_method']).size().reset_index(name='count')
payment_pct = payment_channel.copy()
totals = payment_pct.groupby('store_channel')['count'].transform('sum')
payment_pct['pct'] = (payment_pct['count'] / totals * 100).round(1)
print("Payment method split by channel:")
print(payment_pct.to_string(index=False))
```

**Insight (Diagnostic):** If one channel has a higher return rate, it signals a customer experience
or expectation-mismatch problem. If online has a higher avg order value but lower profit margin,
shipping/handling costs may be cutting into profit. The monthly trend shows whether either channel
is growing or declining over the 18-month window.

## Q6 — Are We Spending Marketing Money Wisely?
**Analytics type: Diagnostic + Prescriptive**

We isolate marketing spend from the expenses dataset and compare it against the revenue
generated in the same month to calculate Return on Marketing Spend (ROMS).
We also check whether promotional transactions (`is_promotion = True`) outperform non-promotional ones.

```python
# Isolate marketing expenses
marketing_expenses = expenses[expenses['category'] == 'Marketing'].copy()
marketing_expenses['month'] = marketing_expenses['date'].dt.to_period('M')

monthly_marketing = marketing_expenses.groupby('month')['amount'].sum().reset_index()
monthly_marketing.columns = ['month', 'marketing_spend']

# Monthly revenue
monthly_revenue = sales.groupby('month')['total_sale'].sum().reset_index()
monthly_revenue.columns = ['month', 'revenue']

# Monthly net profit
monthly_profit = sales.groupby('month')['net_profit'].sum().reset_index()
monthly_profit.columns = ['month', 'net_profit']

# Merge
mkt = monthly_marketing.merge(monthly_revenue, on='month', how='inner')
mkt = mkt.merge(monthly_profit, on='month', how='inner')
mkt['roms'] = (mkt['revenue'] / mkt['marketing_spend']).round(2)  # revenue per $1 spent on marketing
mkt['month_dt'] = mkt['month'].dt.to_timestamp()

print("Monthly Marketing Spend vs Revenue:")
print(mkt[['month','marketing_spend','revenue','net_profit','roms']].to_string(index=False))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Left: marketing spend vs revenue over time
ax2 = axes[0].twinx()
axes[0].bar(mkt['month_dt'], mkt['marketing_spend'], color=ORANGE, alpha=0.6, label='Marketing Spend')
ax2.plot(mkt['month_dt'], mkt['revenue'], color=BLUE, marker='o', label='Revenue')
axes[0].set_title("Monthly Marketing Spend vs Revenue")
axes[0].set_ylabel("Marketing Spend ($)", color=ORANGE)
ax2.set_ylabel("Revenue ($)", color=BLUE)
axes[0].tick_params(axis='x', rotation=45)
axes[0].legend(loc='upper left')
ax2.legend(loc='upper right')

# Right: ROMS over time
axes[1].plot(mkt['month_dt'], mkt['roms'], color=GREEN, marker='o', linewidth=2)
axes[1].axhline(mkt['roms'].mean(), color=RED, linestyle='--', label=f"Avg ROMS: {mkt['roms'].mean():.1f}x")
axes[1].set_title("Return on Marketing Spend (Revenue per $1 Spent)")
axes[1].set_ylabel("ROMS (x)")
axes[1].legend()
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Promotion vs non-promotion performance
promo = sales.groupby('is_promotion').agg(
    transactions     = ('sale_id', 'count'),
    avg_order_value  = ('total_sale', 'mean'),
    avg_net_profit   = ('net_profit', 'mean'),
    avg_units_sold   = ('quantity_sold', 'mean'),
    return_rate      = ('is_returned', 'mean')
).reset_index()

promo['return_rate']     = (promo['return_rate'] * 100).round(1)
promo['avg_order_value'] = promo['avg_order_value'].round(2)
promo['avg_net_profit']  = promo['avg_net_profit'].round(2)
promo['label'] = promo['is_promotion'].map({True: 'Promotion', False: 'No Promotion'})
print("Promotional vs Non-Promotional Transactions:")
print(promo[['label','transactions','avg_order_value','avg_net_profit','avg_units_sold','return_rate']].to_string(index=False))
```

**Insight (Prescriptive):** If ROMS is declining over time, marketing spend is becoming less efficient.
If promotional transactions have a lower avg net profit than non-promotional ones, promotions are
cannibalising margin. StyleSense should direct marketing budget toward months with historically
high ROMS and reduce promotions on already high-demand products.

## Q7 — Which Products Are Frequently Out of Stock?
**Analytics type: Diagnostic + Prescriptive**

We use the near-stockout signal (`stock_level <= reorder_point`) to rank products by stockout frequency
and identify the months they are most vulnerable. This directly informs reorder scheduling.

```python
# Near-stockout events per product
near_stockout = inventory[inventory['stock_level'] <= inventory['reorder_point']].copy()

stockout_summary = near_stockout.groupby(['product_id','product_name','category']).agg(
    near_stockout_days = ('date', 'count'),
    min_stock_reached  = ('stock_level', 'min'),
    avg_stock_at_risk  = ('stock_level', 'mean')
).reset_index()

stockout_summary['avg_stock_at_risk'] = stockout_summary['avg_stock_at_risk'].round(1)
stockout_summary = stockout_summary.sort_values('near_stockout_days', ascending=False)

print("Products Most Frequently Near Stockout:")
print(stockout_summary.to_string(index=False))
```

```python
# Heatmap: near-stockout days by product and month
near_stockout['month'] = near_stockout['date'].dt.month
near_stockout['month_name'] = near_stockout['date'].dt.strftime('%b')

heatmap_data = near_stockout.groupby(['product_name','month']).size().reset_index(name='days_at_risk')
heatmap_pivot = heatmap_data.pivot(index='product_name', columns='month', values='days_at_risk').fillna(0)

# Rename columns to month names
month_names = {1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',
               7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}
heatmap_pivot.columns = [month_names.get(c, c) for c in heatmap_pivot.columns]

plt.figure(figsize=(16, 8))
sns.heatmap(heatmap_pivot, annot=True, fmt='.0f', cmap='Reds', linewidths=0.5,
            cbar_kws={'label': 'Days at Risk'})
plt.title("Near-Stockout Days by Product and Month", fontsize=13, fontweight='bold')
plt.xlabel("Month")
plt.ylabel("Product")
plt.tight_layout()
plt.show()
```

```python
# Stockout risk by category
category_stockout = near_stockout.groupby('category').agg(
    near_stockout_days = ('date', 'count'),
    products_affected  = ('product_name', 'nunique')
).reset_index()
category_stockout['days_per_product'] = (category_stockout['near_stockout_days'] / category_stockout['products_affected']).round(1)
category_stockout = category_stockout.sort_values('near_stockout_days', ascending=False)

print("Stockout Risk by Category:")
print(category_stockout.to_string(index=False))
```

**Insight (Prescriptive):** Products with the highest near-stockout days need larger safety stock buffers
or more frequent reorder cycles. The heatmap identifies the exact months each product is most at risk —
allowing buyers to schedule deliveries to arrive *before* those months.
High-risk categories should be prioritised in the annual procurement plan.

## Q8 — What Is Our True Profit After All Expenses?
**Analytics type: Descriptive + Prescriptive**

`gross_profit` in the sales data accounts for COGS at the transaction level.
But StyleSense also has fixed and variable operating expenses (rent, payroll, marketing, etc.).
True net operating profit = total net profit from sales minus all operating expenses.

```python
# Total net profit from sales (after returns)
total_net_profit_sales = sales['net_profit'].sum()

# Total operating expenses
total_expenses = expenses['amount'].sum()

# True operating profit
true_operating_profit = total_net_profit_sales - total_expenses

print(f"Total Net Profit from Sales (after returns) : ${total_net_profit_sales:>12,.2f}")
print(f"Total Operating Expenses                    : ${total_expenses:>12,.2f}")
print(f"True Operating Profit                       : ${true_operating_profit:>12,.2f}")
print(f"Operating Profit Margin                     : {true_operating_profit / sales['total_sale'].sum() * 100:.1f}%")
```

```python
# Monthly true profit
monthly_sales_profit = sales.groupby('month')['net_profit'].sum().reset_index()
monthly_sales_profit.columns = ['month', 'net_profit']

expenses['month'] = expenses['date'].dt.to_period('M')
monthly_expenses = expenses.groupby('month')['amount'].sum().reset_index()
monthly_expenses.columns = ['month', 'total_expenses']

monthly_pnl = monthly_sales_profit.merge(monthly_expenses, on='month', how='inner')
monthly_pnl['true_profit'] = monthly_pnl['net_profit'] - monthly_pnl['total_expenses']
monthly_pnl['month_dt']    = monthly_pnl['month'].dt.to_timestamp()

print("Monthly P&L Summary:")
print(monthly_pnl[['month','net_profit','total_expenses','true_profit']].to_string(index=False))
```

```python
fig, axes = plt.subplots(2, 1, figsize=(16, 10))

# Top: stacked view of profit vs expenses
axes[0].fill_between(monthly_pnl['month_dt'], monthly_pnl['net_profit'], alpha=0.4, color=GREEN, label='Net Profit (Sales)')
axes[0].fill_between(monthly_pnl['month_dt'], monthly_pnl['total_expenses'], alpha=0.4, color=RED, label='Total Expenses')
axes[0].plot(monthly_pnl['month_dt'], monthly_pnl['net_profit'], color=GREEN, linewidth=2)
axes[0].plot(monthly_pnl['month_dt'], monthly_pnl['total_expenses'], color=RED, linewidth=2)
axes[0].set_title("Monthly Net Profit vs Total Expenses")
axes[0].set_ylabel("Amount ($)")
axes[0].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"${x:,.0f}"))
axes[0].legend()
axes[0].tick_params(axis='x', rotation=45)

# Bottom: true operating profit per month
colors = [GREEN if v >= 0 else RED for v in monthly_pnl['true_profit']]
axes[1].bar(monthly_pnl['month_dt'], monthly_pnl['true_profit'], color=colors, width=20)
axes[1].axhline(0, color='black', linewidth=0.8)
axes[1].set_title("Monthly True Operating Profit (Sales Profit - All Expenses)")
axes[1].set_ylabel("Operating Profit ($)")
axes[1].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"${x:,.0f}"))
axes[1].tick_params(axis='x', rotation=45)

plt.suptitle("StyleSense P&L Overview — Jan 2024 to Jun 2025", fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Expense breakdown: where does the money go?
expense_breakdown = expenses.groupby('category')['amount'].sum().reset_index()
expense_breakdown = expense_breakdown.sort_values('amount', ascending=False)
expense_breakdown['pct'] = (expense_breakdown['amount'] / expense_breakdown['amount'].sum() * 100).round(1)

print("Expense Breakdown:")
print(expense_breakdown.to_string(index=False))

fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(expense_breakdown['amount'], labels=expense_breakdown['category'],
       autopct='%1.1f%%', startangle=140)
ax.set_title("Operating Expense Distribution", fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

**Insight (Prescriptive):** Months where true operating profit turns negative are loss months —
the business is spending more than it earns after all costs.
The expense pie chart reveals which cost categories have the most room for optimisation.
If COGS dominates, product sourcing costs should be renegotiated.
If fixed costs (Rent, Insurance) are a large share of a loss month's expenses,
StyleSense should focus on driving sales volume rather than cutting fixed overhead.
