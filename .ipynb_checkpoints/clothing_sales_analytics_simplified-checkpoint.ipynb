# 👗 StyleSense Retail Analytics
## Turning Sales Data into Smarter Business Decisions

---

### 🏢 Business Background

**StyleSense** is a growing clothing retail company that sells Men's, Women's, and Kids' clothing both in physical stores and online. Like many retail businesses today, StyleSense faces growing competition from large e-commerce platforms and needs to make smarter, data-driven decisions to stay profitable.

---

### 🚨 Business Problem

Despite having consistent sales, StyleSense's management has noticed:

- **Profit margins are shrinking** – expenses are rising but it's unclear if revenue is keeping up
- **Some products keep going out of stock** – causing lost sales and unhappy customers
- **Promotions and discounts are being used** – but no one knows if they're actually helping or just cutting into profit
- **No one knows which customer segments are most valuable** – marketing spend may be wasted

---

### 🎯 Project Objective

This project aims to analyze 2 months of sales, customer, inventory, and expense data to:

1. **Understand** what has been selling and who is buying (Descriptive)
2. **Diagnose** why performance is the way it is (Diagnostic)
3. **Predict** future revenue trends (Predictive)
4. **Recommend** clear actions management can take to improve ROI (Prescriptive)

---

### 📂 Datasets Used

| File | What it Contains |
|------|------------------|
| `clothing_sales.csv` | Every sale transaction with product, store, date, price, discount |
| `clothing_customers.csv` | Customer profiles – age, gender, purchase history |
| `clothing_expenses.csv` | Daily company expenses by category |
| `clothing_inventory.csv` | Daily stock levels per product |

---
## STEP 1 — Import Tools & Load Data

Before we do anything, we bring in the Python libraries (tools) we need:
- **pandas** — for working with tables of data
- **matplotlib** — for drawing charts
- **numpy** — for math calculations

```python
# Import all necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')  # Hide unnecessary warning messages

print('✅ All libraries loaded successfully!')
```

```python
# Load all four datasets
sales     = pd.read_csv('clothing_sales.csv')
customers = pd.read_csv('clothing_customers.csv')
expenses  = pd.read_csv('clothing_expenses.csv')
inventory = pd.read_csv('clothing_inventory.csv')

print('✅ All datasets loaded!')
print('Sales rows:    ', len(sales))
print('Customers rows:', len(customers))
print('Expenses rows: ', len(expenses))
print('Inventory rows:', len(inventory))
```

---
## STEP 2 — Data Cleaning

🧹 **Why clean data?**
Real-world data is messy. It can have missing values, wrong data types, duplicates, or formatting issues. If we skip this step, our analysis will give us wrong answers — like getting directions from a map with wrong street names.

We will check for:
- **Missing values** (empty cells)
- **Duplicate rows** (same transaction recorded twice)
- **Correct data types** (dates should be dates, numbers should be numbers)
- **Logical errors** (e.g. negative quantities)

```python
# Preview the sales data — look at the first 5 rows
print('====== SALES DATA ======')
print(sales.head())
print('\nShape:', sales.shape)  # Rows and columns
```

```python
# Check column data types — are they stored correctly?
print('--- Data Types ---')
print(sales.dtypes)

# Check for missing values in each column
print('\n--- Missing Values ---')
print(sales.isnull().sum())
```

```python
# Fix: convert the 'date' columns from text to proper date format
# Without this, Python treats dates as plain text and can't do date math
sales['date']                    = pd.to_datetime(sales['date'])
customers['first_purchase_date'] = pd.to_datetime(customers['first_purchase_date'])
expenses['date']                 = pd.to_datetime(expenses['date'])
inventory['date']                = pd.to_datetime(inventory['date'])

print('✅ Date columns converted to datetime format')
```

```python
# Check for duplicate rows in sales data
duplicates = sales.duplicated().sum()
print('Duplicate rows found:', duplicates)

if duplicates > 0:
    sales = sales.drop_duplicates()
    print('✅ Duplicates removed')
else:
    print('✅ No duplicates — data is clean!')
```

```python
# Check for logical errors — quantities and prices must always be positive numbers
negative_qty   = (sales['quantity_sold'] < 0).sum()
negative_price = (sales['price_per_unit'] < 0).sum()
negative_total = (sales['total_sale'] < 0).sum()

print('Rows with negative quantity:', negative_qty)
print('Rows with negative price:   ', negative_price)
print('Rows with negative total:   ', negative_total)

if negative_qty == 0 and negative_price == 0 and negative_total == 0:
    print('✅ No logical errors found!')
```

```python
# Check the customers dataset
print('====== CUSTOMERS DATA ======')
print(customers.head())
print('\nMissing values:')
print(customers.isnull().sum())
print('\nDuplicates:', customers.duplicated().sum())
```

```python
# Check the expenses dataset
print('====== EXPENSES DATA ======')
print(expenses.head())
print('\nMissing values:')
print(expenses.isnull().sum())
print('\nDuplicates:', expenses.duplicated().sum())
```

```python
# Check the inventory dataset
print('====== INVENTORY DATA ======')
print(inventory.head())
print('\nMissing values:')
print(inventory.isnull().sum())
print('\nUnique products:', inventory['product_id'].nunique())
```

```python
# Add helpful new columns to the sales table for easier analysis
sales['month']    = sales['date'].dt.month_name()       # e.g. 'January'
sales['month_num'] = sales['date'].dt.month             # e.g. 1
sales['week']     = sales['date'].dt.isocalendar().week.astype(int)  # Week number
sales['day_name'] = sales['date'].dt.day_name()         # e.g. 'Monday'

# Calculate cost and profit for each sale
# We assume the company buys items at 50% of the selling price (a common retail assumption)
sales['cost']   = sales['quantity_sold'] * sales['price_per_unit'] * 0.50
sales['profit'] = sales['total_sale'] - sales['cost']

print('✅ New columns added: month, week, day_name, cost, profit')
print(sales[['date', 'total_sale', 'cost', 'profit']].head())
```

```python
# Final cleaning summary
print('======= CLEANING SUMMARY =======')
print('Sales records ready for analysis:', len(sales))
print('Customer records:                ', len(customers))
print('Expense records:                 ', len(expenses))
print('Inventory records:               ', len(inventory))
print('Date range:', sales['date'].min().date(), 'to', sales['date'].max().date())
print('\n✅ Data is clean and ready for analysis!')
```

---
## STEP 3 — Descriptive Analysis
### 📊 What happened? A summary of our business performance.

Descriptive analysis answers: **"What does our data show?"**
Think of this like reading a company's report card — we're summarizing the facts before drawing any conclusions.

```python
# Calculate the key business numbers
total_revenue  = sales['total_sale'].sum()
total_profit   = sales['profit'].sum()
total_expenses = expenses['amount'].sum()
total_units    = sales['quantity_sold'].sum()
net_income     = total_profit - total_expenses
profit_margin  = (total_profit / total_revenue) * 100

print('============================================')
print('       OVERALL BUSINESS SCORECARD')
print('============================================')
print('Total Revenue:  $', round(total_revenue, 2))
print('Gross Profit:   $', round(total_profit, 2))
print('Total Expenses: $', round(total_expenses, 2))
print('Net Income:     $', round(net_income, 2))
print('Total Units Sold:', total_units)
print('Profit Margin:  ', round(profit_margin, 1), '%')
print('============================================')

# Revenue = all money coming in from sales
# Gross Profit = revenue minus what we paid for the goods
# Net Income = what's left after paying ALL expenses
```

```python
# ── CHART 1: Revenue by Month ──────────────────────────────────
# Group sales by month and add up the totals
monthly_revenue = sales.groupby('month')['total_sale'].sum()
monthly_revenue = monthly_revenue.reindex(['January', 'February'])

# Draw the bar chart
plt.figure(figsize=(7, 5))
plt.bar(['January', 'February'], monthly_revenue.values, color=['#4A90D9', '#E67E22'], width=0.5)

# Add labels on top of each bar
plt.text(0, monthly_revenue['January'] + 30, '$' + str(round(monthly_revenue['January'])), ha='center', fontsize=11)
plt.text(1, monthly_revenue['February'] + 30, '$' + str(round(monthly_revenue['February'])), ha='center', fontsize=11)

plt.title('Monthly Revenue Comparison', fontsize=14)
plt.ylabel('Revenue (USD)')
plt.xlabel('Month')
plt.tight_layout()
plt.show()

# 📌 CHART EXPLANATION:
# This bar chart compares how much money the store made in January vs February.
# A taller bar means more money was earned that month.
# We want to see growth from month to month — that means the business is improving.
```

```python
# ── CHART 2: Sales by Product Category ────────────────────────
# Add up revenue per category
category_sales = sales.groupby('category')['total_sale'].sum().sort_values(ascending=False)

# Get the category names and their values separately — easier to work with
cat_names  = list(category_sales.index)
cat_values = list(category_sales.values)
colors     = ['#3498DB', '#E74C3C', '#2ECC71']

# Left chart: bar chart
plt.figure(figsize=(13, 5))

plt.subplot(1, 2, 1)  # This means: 1 row, 2 columns, and we're on chart #1
plt.bar(cat_names, cat_values, color=colors)
plt.title('Revenue by Category')
plt.ylabel('Total Revenue (USD)')
plt.xlabel('Category')

# Right chart: pie chart
plt.subplot(1, 2, 2)  # Chart #2
plt.pie(cat_values, labels=cat_names, autopct='%1.1f%%', colors=colors, startangle=90)
plt.title('Category Share of Total Revenue')

plt.tight_layout()
plt.show()

# 📌 CHART EXPLANATION:
# The bar chart (left) shows which clothing category made the most money.
# The pie chart (right) shows each category as a slice — the bigger the slice, the bigger the share.
# This tells management which category to stock more of and market harder.
```

```python
# ── CHART 3: Top 5 Best-Selling Products ──────────────────────
# Sum up revenue per product, then keep only the top 5
product_revenue = sales.groupby('product_name')['total_sale'].sum()
product_revenue = product_revenue.sort_values(ascending=False).head(5)

product_names  = list(product_revenue.index)
product_values = list(product_revenue.values)

plt.figure(figsize=(9, 5))
plt.barh(product_names, product_values, color='#5DADE2')  # barh = horizontal bars
plt.title('Top 5 Products by Revenue')
plt.xlabel('Total Revenue (USD)')
plt.gca().invert_yaxis()  # Put the highest bar at the top
plt.tight_layout()
plt.show()

print('\nTop 5 Products:')
print(product_revenue)

# 📌 CHART EXPLANATION:
# A horizontal bar chart is easier to read when names are long.
# The longer the bar, the more revenue that product generated.
# These are the STAR products — always keep them in stock and highlight them in ads.
```

```python
# ── CHART 4: Online vs Physical Store ─────────────────────────
# Group by store type and calculate totals
store_revenue = sales.groupby('store')['total_sale'].sum()
store_avg     = sales.groupby('store')['total_sale'].mean().round(2)

store_names = list(store_revenue.index)
rev_values  = list(store_revenue.values)
avg_values  = list(store_avg.values)
colors2     = ['#E74C3C', '#3498DB']

plt.figure(figsize=(12, 5))

# Left: total revenue
plt.subplot(1, 2, 1)
plt.bar(store_names, rev_values, color=colors2)
plt.title('Total Revenue: Online vs Physical')
plt.ylabel('Revenue (USD)')
plt.xlabel('Store Type')

# Right: average sale value per transaction
plt.subplot(1, 2, 2)
plt.bar(store_names, avg_values, color=colors2)
plt.title('Avg Sale Value: Online vs Physical')
plt.ylabel('Avg Sale ($)')
plt.xlabel('Store Type')

plt.tight_layout()
plt.show()

print('Store Revenue Summary:')
print('Total Revenue:\n', store_revenue)
print('\nAverage Sale:\n', store_avg)

# 📌 CHART EXPLANATION:
# Left: which channel brings in more total money?
# Right: which channel gets more money per single sale?
# A channel with fewer sales but higher average can still be more valuable per customer visit.
```

```python
# ── CHART 5: Customer Age & Gender Profile ─────────────────────
print('CUSTOMER DEMOGRAPHICS SUMMARY:')
print(customers[['age', 'total_purchases']].describe().round(2))

avg_age      = customers['age'].mean()
gender_counts = customers['gender'].value_counts()
gender_labels = ['Male' if g == 'M' else 'Female' for g in gender_counts.index]

plt.figure(figsize=(12, 5))

# Left: Age distribution (histogram)
plt.subplot(1, 2, 1)
plt.hist(customers['age'], bins=8, color='#8E44AD', edgecolor='white')
plt.axvline(avg_age, color='red', linestyle='--', label='Avg Age: ' + str(round(avg_age)))
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Customers')
plt.legend()

# Right: Gender pie chart
plt.subplot(1, 2, 2)
plt.pie(gender_counts.values, labels=gender_labels, autopct='%1.1f%%',
        colors=['#3498DB', '#E91E63'], startangle=90)
plt.title('Gender Distribution of Customers')

plt.tight_layout()
plt.show()

# 📌 CHART EXPLANATION:
# The histogram (left) groups customers by age. Each bar shows how many customers fall in that age range.
# The red dashed line is the average age — most customers cluster around this point.
# The pie chart (right) shows the gender split of our customer base.
# This helps us create marketing materials that speak to the right audience.
```

```python
# ── CHART 6: Expense Breakdown ─────────────────────────────────
expense_by_cat = expenses.groupby('category')['amount'].sum().sort_values(ascending=False)

exp_labels = list(expense_by_cat.index)
exp_values = list(expense_by_cat.values)

plt.figure(figsize=(7, 6))
plt.pie(exp_values, labels=exp_labels, autopct='%1.1f%%', startangle=90)
plt.title('Total Expenses by Category')
plt.tight_layout()
plt.show()

print('\nExpense Totals:')
print(expense_by_cat)

# 📌 CHART EXPLANATION:
# This pie chart breaks down WHERE the company spends its money.
# The biggest slice = the biggest cost. That's the first place to look when trying to cut costs.
# A healthy business spreads expenses without one category completely dominating.
```

## STEP 4 — Diagnostic Analysis
### 🔍 Why did things happen the way they did?

Descriptive told us *what* happened. Diagnostic analysis helps us understand *why* it happened.
We dig deeper — looking at patterns, relationships, and what's driving the numbers.

```python
# ── CHART 7: Impact of Promotions ─────────────────────────────
# Does running a promotion actually increase sales and profit?

promo_avg_sale   = sales.groupby('promotion')['total_sale'].mean().round(2)
promo_avg_profit = sales.groupby('promotion')['profit'].mean().round(2)

promo_labels = ['No Promotion', 'With Promotion']
sale_vals    = list(promo_avg_sale.values)
profit_vals  = list(promo_avg_profit.values)
promo_colors = ['#95A5A6', '#F39C12']

plt.figure(figsize=(12, 5))

# Left: average sale value
plt.subplot(1, 2, 1)
plt.bar(promo_labels, sale_vals, color=promo_colors)
plt.title('Avg Sale Value: With vs Without Promotion')
plt.ylabel('Avg Sale ($)')

# Right: average profit
plt.subplot(1, 2, 2)
plt.bar(promo_labels, profit_vals, color=promo_colors)
plt.title('Avg Profit: With vs Without Promotion')
plt.ylabel('Avg Profit ($)')

plt.tight_layout()
plt.show()

print('Average Sale — No Promo:', sale_vals[0], '| With Promo:', sale_vals[1])
print('Average Profit — No Promo:', profit_vals[0], '| With Promo:', profit_vals[1])

# 📌 CHART EXPLANATION:
# Left chart: Are customers spending MORE when there's a promotion running?
# Right chart: Are we making MORE profit from promoted sales, or are the discounts hurting us?
# If sales go up but profit goes down, the promotion is bringing in customers but costing us money.
```

```python
# ── CHART 8: Revenue by Day of the Week ───────────────────────
# Which day do customers shop the most? This helps with staffing and marketing timing.

day_order   = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daily_sales = sales.groupby('day_name')['total_sale'].sum()
daily_sales = daily_sales.reindex(day_order)  # Make sure days are in the right order

# Color weekends differently from weekdays
bar_colors = ['#E74C3C', '#E74C3C', '#E74C3C', '#E74C3C', '#E74C3C', '#27AE60', '#27AE60']
# First 5 are weekdays (red), last 2 are weekends (green)

plt.figure(figsize=(10, 5))
plt.bar(day_order, daily_sales.values, color=bar_colors)
plt.title('Total Revenue by Day of the Week')
plt.ylabel('Total Revenue (USD)')
plt.xlabel('Day')
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()

# 📌 CHART EXPLANATION:
# Red bars = weekdays. Green bars = weekends.
# The tallest bar is the busiest sales day.
# If weekends dominate, schedule more staff and run promotions on those days.
# If weekdays are stronger, our online channel is likely driving daytime purchases.
```

```python
# ── CHART 9: Discount Level vs Profit ─────────────────────────
# Are big discounts hurting our profit margins?

# Group discounts into 4 buckets
bins   = [-1, 0, 9, 19, 100]
labels = ['No Discount', 'Low (1-9%)', 'Medium (10-19%)', 'High (20%+)']
sales['discount_group'] = pd.cut(sales['discount'], bins=bins, labels=labels)

discount_avg_profit = sales.groupby('discount_group', observed=True)['profit'].mean().round(2)

disc_labels = list(discount_avg_profit.index)
disc_values = list(discount_avg_profit.values)
disc_colors = ['#27AE60', '#F1C40F', '#E67E22', '#E74C3C']

plt.figure(figsize=(8, 5))
plt.bar(disc_labels, disc_values, color=disc_colors)
plt.title('Average Profit by Discount Level')
plt.ylabel('Average Profit ($)')
plt.xlabel('Discount Level')
plt.xticks(rotation=10)
plt.tight_layout()
plt.show()

print('Average Profit by Discount Level:')
print(discount_avg_profit)

# 📌 CHART EXPLANATION:
# Green bar = no discount = most profit per sale.
# As we move right to bigger discounts, the bars should ideally stay as tall as possible.
# If the red (High Discount) bar is very short, large discounts are eating into profit heavily.
# This tells management how much discounting is too much.
```

```python
# ── CHART 10: Stockout Analysis ────────────────────────────────
# A stockout = stock level is 0 → we can't sell → lost revenue

# Filter for days where stock hit zero, then count how many days per product
zero_stock = inventory[inventory['stock_level'] == 0]
stockout_counts = zero_stock.groupby('product_id').size().reset_index()
stockout_counts.columns = ['product_id', 'days_out_of_stock']
stockout_counts = stockout_counts.sort_values('days_out_of_stock', ascending=False)

print('Days Out of Stock per Product:')
print(stockout_counts)

plt.figure(figsize=(10, 5))
plt.bar(stockout_counts['product_id'], stockout_counts['days_out_of_stock'], color='#E74C3C')
plt.title('Days Out of Stock per Product')
plt.ylabel('Number of Days with Zero Stock')
plt.xlabel('Product ID')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 📌 CHART EXPLANATION:
# Each red bar shows how many days a product had ZERO stock.
# The taller the bar, the more selling days were lost because the item was unavailable.
# Products with the tallest bars need a better restocking plan — these are hurting our revenue.
```

```python
# ── CHART 11: Daily Revenue vs Expenses ───────────────────────
# Are we earning more than we spend each day?

daily_rev = sales.groupby('date')['total_sale'].sum().reset_index()
daily_exp = expenses.groupby('date')['amount'].sum().reset_index()

# Combine both into one table, fill missing dates with 0
combined = pd.merge(daily_rev, daily_exp, on='date', how='outer').fillna(0)
combined.columns = ['date', 'revenue', 'expenses']
combined = combined.sort_values('date')

plt.figure(figsize=(12, 5))
plt.plot(combined['date'], combined['revenue'],  color='#27AE60', linewidth=2, label='Revenue')
plt.plot(combined['date'], combined['expenses'], color='#E74C3C', linewidth=2, linestyle='--', label='Expenses')
plt.title('Daily Revenue vs Expenses Over Time')
plt.ylabel('Amount ($)')
plt.xlabel('Date')
plt.legend()
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

# 📌 CHART EXPLANATION:
# Green solid line = money coming IN (revenue) each day.
# Red dashed line = money going OUT (expenses) each day.
# When the green line is ABOVE the red line, we're making profit that day.
# When the red line is ABOVE the green line, we're losing money that day — a warning sign.
```

## STEP 5 — Predictive Analysis
### 🔮 What is likely to happen next?

Predictive analysis uses patterns from past data to forecast the future.
We use **Linear Regression** — a simple method that draws a straight trend line through our data and extends it forward to estimate future values.

Think of it like this: if revenue was $200 in Week 1 and $250 in Week 2, the trend line predicts roughly $300 in Week 3.

```python
# Import the tool that builds our prediction model
from sklearn.linear_model import LinearRegression

# Step 1: Get total revenue per week
weekly_sales = sales.groupby('week')['total_sale'].sum().reset_index()
weekly_sales.columns = ['week', 'revenue']

print('Weekly Revenue Data:')
print(weekly_sales)
```

```python
# Step 2: Set up the model inputs
# X = week number (what we know — the input)
# y = revenue for that week (what we want to predict — the output)
X = weekly_sales[['week']]  # Double brackets needed for sklearn
y = weekly_sales['revenue']

# Step 3: Create and train the model
model = LinearRegression()
model.fit(X, y)  # The model learns the trend from our data

# Step 4: Make future week numbers and predict revenue for them
last_week     = weekly_sales['week'].max()
future_weeks  = pd.DataFrame({'week': [last_week+1, last_week+2, last_week+3, last_week+4]})
future_revenue = model.predict(future_weeks)

print('PREDICTED REVENUE FOR NEXT 4 WEEKS:')
for i in range(4):
    print('  Week', future_weeks['week'].iloc[i], ': $', round(future_revenue[i], 2))
```

```python
# ── CHART 12: Actual vs Predicted Revenue ─────────────────────
# Get what the model thinks happened in the existing weeks
predicted_existing = model.predict(X)

plt.figure(figsize=(11, 5))

# Plot actual weekly revenue as dots
plt.scatter(weekly_sales['week'], weekly_sales['revenue'], color='#2C3E50', s=80, label='Actual Revenue', zorder=5)

# Plot the trend line through existing data
plt.plot(weekly_sales['week'], predicted_existing, color='#3498DB', linewidth=2, label='Trend Line')

# Plot predicted future revenue as red star markers
plt.scatter(future_weeks['week'], future_revenue, color='#E74C3C', s=150, marker='*', label='Predicted (Next 4 Weeks)', zorder=5)
plt.plot(future_weeks['week'], future_revenue, color='#E74C3C', linewidth=2, linestyle='--')

# Draw a vertical line to show where forecast begins
plt.axvline(x=last_week + 0.5, color='gray', linestyle=':', linewidth=1.5, label='Forecast Starts Here')

plt.title('Weekly Revenue: Actual vs Predicted (Next 4 Weeks)')
plt.xlabel('Week Number')
plt.ylabel('Revenue ($)')
plt.legend()
plt.tight_layout()
plt.show()

# 📌 CHART EXPLANATION:
# Black dots = what actually happened each week.
# Blue line = the trend that runs through those dots.
# Red stars = what our model PREDICTS for the next 4 weeks based on that trend.
# Dotted vertical line = where our actual data ends and the forecast begins.
# A rising trend means the business is growing — a falling trend needs attention.
```

```python
# ── Model Accuracy Check ──────────────────────────────────────
# R² score = how well the model fits the data (1.0 = perfect, 0 = useless)
# MAE = on average, how many dollars off is our prediction?

from sklearn.metrics import r2_score, mean_absolute_error

r2  = r2_score(y, predicted_existing)
mae = mean_absolute_error(y, predicted_existing)

print('Model R² Score:', round(r2, 3))
print('Mean Absolute Error: $', round(mae, 2))

if r2 >= 0.7:
    print('\n✅ Strong fit — predictions are reasonably reliable.')
elif r2 >= 0.4:
    print('\n⚠️  Moderate fit — use as a directional guide, not exact forecast.')
else:
    print('\n❌ Weak fit — predictions are rough estimates. More data will help.')

# 📌 INTERPRETATION:
# If R² = 0.85, the model explains 85% of the revenue changes week-to-week — quite good!
# MAE tells us the average dollar gap between our prediction and reality.
```

```python
# ── CHART 13: Category Growth Rate ────────────────────────────
# Which clothing category is growing the fastest?

categories     = list(sales['category'].unique())
growth_rates   = []
growth_cats    = []

for cat in categories:
    # Filter sales for this category only
    cat_data = sales[sales['category'] == cat].groupby('week')['total_sale'].sum().reset_index()
    cat_data.columns = ['week', 'revenue']

    if len(cat_data) >= 2:  # Need at least 2 weeks to draw a trend line
        X_cat = cat_data[['week']]
        y_cat = cat_data['revenue']
        m = LinearRegression()
        m.fit(X_cat, y_cat)
        slope = m.coef_[0]  # Positive = growing, Negative = declining
        growth_rates.append(slope)
        growth_cats.append(cat)

# Color bars: green if growing, red if shrinking
bar_colors_growth = ['#27AE60' if r > 0 else '#E74C3C' for r in growth_rates]

plt.figure(figsize=(7, 5))
plt.bar(growth_cats, growth_rates, color=bar_colors_growth)
plt.axhline(0, color='black', linewidth=0.8)  # Zero line for reference
plt.title('Predicted Weekly Revenue Growth Rate by Category')
plt.ylabel('Revenue Change Per Week ($)')
plt.xlabel('Category')
plt.tight_layout()
plt.show()

print('\nGrowth Rates by Category:')
for cat, rate in zip(growth_cats, growth_rates):
    print(' ', cat, ':', round(rate, 2), '$/week')

# 📌 CHART EXPLANATION:
# Green bars = the category is growing week over week — invest more here.
# Red bars = the category is declining — investigate why and take action.
# The taller the green bar, the faster the growth.
```

## STEP 6 — Prescriptive Analysis
### 💡 What should we DO about it?

Prescriptive analysis is the most valuable — it turns all our findings into **concrete, actionable recommendations**.
Instead of just saying "sales are low," we say: *"Here's exactly what to do to fix it."*

```python
# ── ACTION 1: Product Priority Scoring ────────────────────────
# Which products deserve the most investment?
# We score each product by combining revenue AND profit margin

# Get revenue and profit per product
prod_revenue = sales.groupby('product_name')['total_sale'].sum()
prod_profit  = sales.groupby('product_name')['profit'].sum()

product_scores = pd.DataFrame({'total_revenue': prod_revenue, 'total_profit': prod_profit})
product_scores['profit_margin'] = product_scores['total_profit'] / product_scores['total_revenue']

# Normalize to a 0-1 scale so we can combine them fairly
rev_min = product_scores['total_revenue'].min()
rev_max = product_scores['total_revenue'].max()
mar_min = product_scores['profit_margin'].min()
mar_max = product_scores['profit_margin'].max()

product_scores['rev_score']    = (product_scores['total_revenue'] - rev_min) / (rev_max - rev_min)
product_scores['margin_score'] = (product_scores['profit_margin'] - mar_min) / (mar_max - mar_min)

# 60% weight on revenue, 40% on margin
product_scores['priority_score'] = (0.6 * product_scores['rev_score']) + (0.4 * product_scores['margin_score'])
product_scores['priority_score'] = product_scores['priority_score'].round(3)
product_scores = product_scores.sort_values('priority_score', ascending=False)

top6 = product_scores.head(6).reset_index()

print('TOP PRODUCTS TO INVEST IN:')
print(top6[['product_name', 'total_revenue', 'profit_margin', 'priority_score']])

plt.figure(figsize=(9, 5))
plt.barh(top6['product_name'], top6['priority_score'], color='#5DADE2')
plt.title('Product Priority Score (Higher = Invest More)')
plt.xlabel('Priority Score (0 = lowest, 1 = highest)')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

# 📌 CHART EXPLANATION:
# Products at the TOP scored highest on BOTH revenue AND profit margin.
# These are your STAR products — always keep them fully stocked and run ads on them.
# Products at the bottom of the list need improvement or reconsideration.
```

```python
# ── ACTION 2: Smart Promotion Strategy ────────────────────────
# Which discount level gives us the best return on investment?

disc_revenue = sales.groupby('discount_group', observed=True)['total_sale'].sum()
disc_profit  = sales.groupby('discount_group', observed=True)['profit'].sum()
disc_roi     = (disc_profit / disc_revenue * 100).round(1)

disc_labels = list(disc_roi.index)
disc_vals   = list(disc_roi.values)
avg_roi     = disc_roi.mean()
disc_colors = ['#27AE60', '#F1C40F', '#E67E22', '#E74C3C']

plt.figure(figsize=(8, 5))
plt.bar(disc_labels, disc_vals, color=disc_colors)
plt.axhline(avg_roi, color='black', linestyle='--', label='Average ROI: ' + str(round(avg_roi, 1)) + '%')
plt.title('Profit ROI % by Discount Level')
plt.ylabel('Profit as % of Revenue')
plt.xlabel('Discount Level')
plt.xticks(rotation=10)
plt.legend()
plt.tight_layout()
plt.show()

print('ROI by Discount Level:')
print(disc_roi)

# 📌 CHART EXPLANATION:
# Green = most profitable (no or low discounts). Red = biggest discounts = lowest profit.
# The dashed black line is the average ROI across all discount types.
# Recommendation: Keep discounts in the range that stays ABOVE the average ROI line.
```

```python
# ── ACTION 3: Inventory Reorder Alerts ────────────────────────
# Which products need to be restocked RIGHT NOW?

# Get the most recent stock level for each product
latest_stock = inventory.sort_values('date').groupby('product_id').last().reset_index()
latest_stock = latest_stock[['product_id', 'stock_level', 'reorder_point', 'supplier']]
latest_stock['needs_reorder'] = latest_stock['stock_level'] <= latest_stock['reorder_point']

reorder_now = latest_stock[latest_stock['needs_reorder']].sort_values('stock_level')

print('⚠️  PRODUCTS THAT NEED IMMEDIATE REORDERING:', len(reorder_now))
print(reorder_now[['product_id', 'stock_level', 'reorder_point', 'supplier']].to_string(index=False))

# Assign bar colors: red if needs reorder, green if stock is fine
stock_colors = []
for needs in latest_stock['needs_reorder']:
    if needs:
        stock_colors.append('#E74C3C')  # Red
    else:
        stock_colors.append('#27AE60')  # Green

avg_reorder = latest_stock['reorder_point'].mean()

plt.figure(figsize=(12, 5))
plt.bar(latest_stock['product_id'], latest_stock['stock_level'], color=stock_colors)
plt.axhline(avg_reorder, color='orange', linestyle='--', linewidth=2, label='Avg Reorder Point')
plt.title('Current Stock Levels (Red = Reorder Needed)')
plt.ylabel('Units in Stock')
plt.xlabel('Product ID')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# 📌 CHART EXPLANATION:
# Red bars = products that have fallen AT or BELOW the safe restocking threshold.
# Green bars = products with healthy stock levels.
# Orange dashed line = the average reorder trigger point.
# Any red bar below the orange line = contact the supplier immediately.
```

```python
# ── ACTION 4: Identify VIP Customers to Retain ────────────────
# Our top 20% customers likely drive a huge share of revenue.
# Keeping them happy is more cost-effective than finding new customers.

# Calculate total spending and number of orders per customer
cust_spending = sales.groupby('customer_id')['total_sale'].sum().reset_index()
cust_spending.columns = ['customer_id', 'total_spent']
cust_spending = cust_spending.sort_values('total_spent', ascending=False).reset_index(drop=True)

# Label top 20% as VIP
top_cutoff = int(len(cust_spending) * 0.2)
cust_spending['segment'] = 'Regular'
cust_spending.loc[:top_cutoff, 'segment'] = 'VIP'

vip_total     = cust_spending[cust_spending['segment'] == 'VIP']['total_spent'].sum()
regular_total = cust_spending[cust_spending['segment'] == 'Regular']['total_spent'].sum()
grand_total   = vip_total + regular_total

print('Total customers:', len(cust_spending))
print('VIP customers (top 20%):', top_cutoff)
print('VIP revenue: $', round(vip_total, 2))
print('VIP share of total revenue:', round(vip_total / grand_total * 100, 1), '%')

plt.figure(figsize=(6, 6))
plt.pie([vip_total, regular_total],
        labels=['VIP (Top 20%)', 'Regular'],
        autopct='%1.1f%%',
        colors=['#F39C12', '#3498DB'],
        startangle=90,
        explode=(0.05, 0))
plt.title('Revenue: VIP vs Regular Customers')
plt.tight_layout()
plt.show()

# 📌 CHART EXPLANATION:
# The orange slice = revenue from VIP customers (top 20% by spending).
# The blue slice = everyone else.
# If orange is a large slice, a small group of customers is very important to the business.
# Recommendation: Create a loyalty program or exclusive offers just for VIP customers.
```
