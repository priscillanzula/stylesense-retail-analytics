# 👗 StyleSense Retail Analytics
## Turning 18 Months of Sales Data into Smarter Business Decisions

---

### 🏢 Business Background

**StyleSense** is a growing clothing retail company that sells Men's, Women's, and Kids' clothing across **Physical stores and Online channels**. Like many retail businesses today, StyleSense faces growing competition from large e-commerce platforms and needs to make smarter, data-driven decisions to stay profitable.

---

### 🚨 Business Problem

Despite consistent sales, StyleSense's management has noticed:

- **Profit margins are unclear** – cost of goods, payroll, rent, and marketing all eat into revenue
- **Seasonality is not being planned for** – winter and summer products go out of stock or sit unsold
- **Promotions and discounts are used freely** – but no one knows if they're actually generating profit
- **Customer segments are not being leveraged** – VIP customers may be leaving due to lack of retention programs
- **Online vs physical performance is unknown** – marketing spend may be mis-allocated

---

### 🎯 Project Objective

This project analyzes **18 months** (Jan 2024 – Jun 2025) of sales, customer, inventory, and expense data to:

1. **Understand** what has been selling and who is buying *(Descriptive)*
2. **Diagnose** why performance is the way it is *(Diagnostic)*
3. **Predict** future revenue trends *(Predictive)*
4. **Recommend** clear actions management can take to improve ROI *(Prescriptive)*

---

### 📂 Datasets Used

| File | Rows | What it Contains |
|------|------|-----------------|
| `clothing_sales.csv` | ~10,000 | Every transaction: product, channel, price, discount, profit |
| `clothing_customers.csv` | 200 | Customer profiles: demographics, segments, purchase history |
| `clothing_expenses.csv` | ~1,300 | Business costs by category (rent, payroll, COGS, marketing) |
| `clothing_inventory.csv` | ~13,600 | Daily stock levels per product with reorder triggers |
| `clothing_products.csv` | 25 | Product master list with base price and cost |

## STEP 1 — Import Libraries & Load Data

Before we do anything, we bring in the Python libraries we need:
- **pandas** — for working with tables of data
- **numpy** — for math calculations
- **matplotlib / seaborn** — for charts and visualisations
- **sklearn** — for predictive modelling


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error
import warnings

warnings.filterwarnings('ignore')

# Chart styling
sns.set_theme(style="whitegrid")
plt.rcParams['figure.dpi'] = 100
plt.rcParams['font.size'] = 10

BLUE   = '#3498DB'
GREEN  = '#27AE60'
RED    = '#E74C3C'
ORANGE = '#E67E22'
PURPLE = '#8E44AD'
GOLD   = '#F39C12'
TEAL   = '#1ABC9C'
GREY   = '#95A5A6'

print('✅ All libraries loaded successfully!')
```

```python
# Load all five datasets
sales     = pd.read_csv('clothing_sales.csv')
customers = pd.read_csv('clothing_customers.csv')
expenses  = pd.read_csv('clothing_expenses.csv')
inventory = pd.read_csv('clothing_inventory.csv')
products  = pd.read_csv('clothing_products.csv')

print('✅ All datasets loaded!')
print(f'  Sales rows:      {len(sales):,}')
print(f'  Customer rows:   {len(customers):,}')
print(f'  Expense rows:    {len(expenses):,}')
print(f'  Inventory rows:  {len(inventory):,}')
print(f'  Product rows:    {len(products):,}')
```

---
## STEP 2 — Data Cleaning

🧹 **Why clean data?**  
Real-world data is messy. It can have missing values, wrong data types, duplicates, or formatting inconsistencies.
If we skip this step, our analysis will give us wrong or misleading answers.

We check for and fix:
- **Missing values** (null / empty cells)
- **Duplicate rows** (same transaction recorded twice)
- **Incorrect data types** (dates stored as text, booleans as strings)
- **Inconsistent formatting** (mixed case in category names)
- **Logical errors** (impossible ages, negative quantities)


```python
# ─── 2.1 Initial data preview ───────────────────────────────────────────────
print('SALES — first 3 rows:')
print(sales.head(3).to_string())
print(f'\nShape: {sales.shape}')
```

```python
# ─── 2.2 Data types and missing values — SALES ──────────────────────────────
print('--- SALES Data Types ---')
print(sales.dtypes)
print('\n--- Missing Values ---')
missing = sales.isnull().sum()
print(missing[missing > 0])
```

```python
# ─── 2.3 Convert all date columns to datetime ───────────────────────────────
date_cols = {
    'sales':     'date',
    'customers': ['first_purchase_date', 'last_purchase_date'],
    'expenses':  'date',
    'inventory': 'date',
}

sales['date']                     = pd.to_datetime(sales['date'])
customers['first_purchase_date']  = pd.to_datetime(customers['first_purchase_date'])
customers['last_purchase_date']   = pd.to_datetime(customers['last_purchase_date'])
expenses['date']                  = pd.to_datetime(expenses['date'])
inventory['date']                 = pd.to_datetime(inventory['date'])

# Convert boolean columns that may have loaded as strings
for col in ['is_promotion', 'is_returned']:
    if sales[col].dtype == object:
        sales[col] = sales[col].str.strip().str.lower() == 'true'

print('✅ Date and boolean columns converted correctly')
print(sales[['date', 'is_promotion', 'is_returned']].dtypes)
```

```python
# ─── 2.4 Fix inconsistent category casing ───────────────────────────────────
# About 2% of category values were lowercase (simulated data entry error)
print('Unique category values BEFORE fix:', sales['category'].unique())

sales['category']     = sales['category'].str.strip().str.title()
sales['sub_category'] = sales['sub_category'].str.strip().str.title()

print('Unique category values AFTER fix: ', sales['category'].unique())
print('✅ Category casing standardised')
```

```python
# ─── 2.5 Handle duplicates ──────────────────────────────────────────────────
dupes = sales.duplicated().sum()
print(f'Duplicate rows found: {dupes}')

if dupes > 0:
    sales = sales.drop_duplicates().reset_index(drop=True)
    print(f'✅ Duplicates removed — {len(sales):,} rows remain')
else:
    print('✅ No duplicates found')
```

```python
# ─── 2.6 Investigate and fix missing values ─────────────────────────────────

# Sales: ~1% are guest checkouts with no customer_id — fill as 'GUEST'
null_cust = sales['customer_id'].isnull().sum()
print(f'Rows with missing customer_id (guest checkouts): {null_cust}')
sales['customer_id'] = sales['customer_id'].fillna('GUEST')

# Sales: a few missing payment_method — fill with 'Unknown'
null_pay = sales['payment_method'].isnull().sum()
print(f'Rows with missing payment_method: {null_pay}')
sales['payment_method'] = sales['payment_method'].fillna('Unknown')

# Customers: outlier ages (0 and 145 are impossible — data entry errors)
print(f'\nCustomer age outliers (age < 15 or age > 100):')
age_outliers = customers[(customers['age'] < 15) | (customers['age'] > 100)]
print(age_outliers[['customer_id', 'name', 'age']])
# Replace with median age
median_age = customers[(customers['age'] >= 15) & (customers['age'] <= 100)]['age'].median()
customers.loc[(customers['age'] < 15) | (customers['age'] > 100), 'age'] = int(median_age)
print(f'✅ Age outliers replaced with median age: {int(median_age)}')

# Inventory: null stock_value — recalculate from stock_level × unit_cost
null_sv = inventory['stock_value'].isnull().sum()
print(f'\nNull stock_value entries: {null_sv}')
inventory['stock_value'] = inventory['stock_level'] * inventory['unit_cost']
print('✅ stock_value recalculated from stock_level × unit_cost')
```

```python
# ─── 2.7 Logical checks ─────────────────────────────────────────────────────
neg_qty    = (sales['quantity_sold'] < 0).sum()
neg_price  = (sales['unit_price'] < 0).sum()
neg_total  = (sales['total_sale'] < 0).sum()
neg_profit = (sales['gross_profit'] < 0).sum()

print(f'Negative quantity_sold:  {neg_qty}')
print(f'Negative unit_price:     {neg_price}')
print(f'Negative total_sale:     {neg_total}')
print(f'Negative gross_profit:   {neg_profit}  ← expected (high discounts)')
print('\n✅ No logical errors in quantity or price fields')
```

```python
# ─── 2.8 Feature engineering — add useful time columns ──────────────────────
sales['year']       = sales['date'].dt.year
sales['month_num']  = sales['date'].dt.month
sales['month_name'] = sales['date'].dt.strftime('%b')          # 'Jan', 'Feb' ...
sales['year_month'] = sales['date'].dt.to_period('M').astype(str)  # '2024-01'
sales['day_name']   = sales['date'].dt.day_name()
sales['week']       = sales['date'].dt.isocalendar().week.astype(int)
sales['is_weekend'] = sales['date'].dt.weekday >= 5
sales['quarter']    = sales['date'].dt.quarter

print('✅ New time columns added')
print(sales[['date', 'year', 'month_name', 'year_month', 'day_name', 'quarter', 'is_weekend']].head(3))
```

```python
# ─── 2.9 Cleaning summary ───────────────────────────────────────────────────
print('=' * 50)
print('       DATA CLEANING COMPLETE — SUMMARY')
print('=' * 50)
print(f'  Sales records ready:    {len(sales):,}')
print(f'  Customer records:       {len(customers):,}')
print(f'  Expense records:        {len(expenses):,}')
print(f'  Inventory records:      {len(inventory):,}')
print(f'  Date range: {sales["date"].min().date()} → {sales["date"].max().date()}')
print(f'  Remaining nulls in sales: {sales.isnull().sum().sum()}')
print('\n✅ Data is clean and ready for analysis!')
```

---
## STEP 3 — Descriptive Analysis
### 📊 What happened? A summary of our business performance.

Descriptive analysis answers: **"What does our data show?"**  
Think of this like reading a company's annual report — we're summarizing the facts before drawing conclusions.


```python
# ─── 3.1 Overall Business Scorecard ────────────────────────────────────────
total_revenue    = sales['total_sale'].sum()
total_gross_prof = sales['gross_profit'].sum()
total_expenses   = expenses['amount'].sum()
total_units      = sales['quantity_sold'].sum()
total_orders     = len(sales)
gross_margin_pct = (total_gross_prof / total_revenue) * 100

# Net income = gross profit minus all non-COGS expenses
non_cogs_exp = expenses[expenses['category'] != 'Cost of Goods Sold']['amount'].sum()
net_income   = total_gross_prof - non_cogs_exp

print('=' * 52)
print('         STYLESENSE — BUSINESS SCORECARD')
print('        Jan 2024 – Jun 2025 (18 months)')
print('=' * 52)
print(f'  Total Revenue:          ${total_revenue:>12,.2f}')
print(f'  Gross Profit:           ${total_gross_prof:>12,.2f}')
print(f'  Gross Margin %:         {gross_margin_pct:>11.1f}%')
print(f'  Operating Expenses:     ${non_cogs_exp:>12,.2f}')
print(f'  Net Income:             ${net_income:>12,.2f}')
print(f'  Total Orders:           {total_orders:>12,}')
print(f'  Total Units Sold:       {total_units:>12,}')
print(f'  Avg Order Value:        ${total_revenue/total_orders:>12.2f}')
print('=' * 52)
```

```python
# ─── CHART 1: Monthly Revenue Trend ────────────────────────────────────────
# Group sales by year-month and sum revenue
monthly_rev = (
    sales.groupby('year_month')['total_sale']
    .sum()
    .reset_index()
)
monthly_rev['period'] = pd.to_datetime(monthly_rev['year_month'])
monthly_rev = monthly_rev.sort_values('period')

fig, ax = plt.subplots(figsize=(14, 5))
bars = ax.bar(monthly_rev['year_month'], monthly_rev['total_sale'],
              color=BLUE, alpha=0.85, width=0.7)

# Add a trend line
x_num = np.arange(len(monthly_rev))
z = np.polyfit(x_num, monthly_rev['total_sale'], 1)
p = np.poly1d(z)
ax.plot(monthly_rev['year_month'], p(x_num), color=RED, linewidth=2,
        linestyle='--', label='Revenue Trend')

ax.set_title('Monthly Revenue Trend — Jan 2024 to Jun 2025', fontsize=13, fontweight='bold')
ax.set_ylabel('Revenue (USD)')
ax.set_xlabel('Month')
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
ax.tick_params(axis='x', rotation=45)
ax.legend()
plt.tight_layout()
plt.show()

print('\n📌 The red dashed line shows the overall revenue trajectory. Rising = healthy growth.')
print(f'   Peak month: {monthly_rev.loc[monthly_rev["total_sale"].idxmax(), "year_month"]} (${monthly_rev["total_sale"].max():,.0f})')
print(f'   Lowest month: {monthly_rev.loc[monthly_rev["total_sale"].idxmin(), "year_month"]} (${monthly_rev["total_sale"].min():,.0f})')
```

```python
# ─── CHART 2: Revenue & Gross Profit by Category ───────────────────────────
# Exclude 'All' (accessories) from primary category view — include separately
cat_summary = (
    sales.groupby('category')
    .agg(revenue=('total_sale', 'sum'), profit=('gross_profit', 'sum'))
    .reset_index()
    .sort_values('revenue', ascending=False)
)
cat_summary['margin_pct'] = (cat_summary['profit'] / cat_summary['revenue'] * 100).round(1)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Bar: revenue
axes[0].bar(cat_summary['category'], cat_summary['revenue'],
            color=[BLUE, RED, GREEN, GOLD, PURPLE])
axes[0].set_title('Revenue by Category')
axes[0].set_ylabel('Revenue ($)')
axes[0].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
axes[0].tick_params(axis='x', rotation=15)

# Bar: gross profit
axes[1].bar(cat_summary['category'], cat_summary['profit'],
            color=[BLUE, RED, GREEN, GOLD, PURPLE])
axes[1].set_title('Gross Profit by Category')
axes[1].set_ylabel('Gross Profit ($)')
axes[1].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
axes[1].tick_params(axis='x', rotation=15)

# Pie: revenue share
axes[2].pie(cat_summary['revenue'], labels=cat_summary['category'],
            autopct='%1.1f%%', startangle=90,
            colors=[BLUE, RED, GREEN, GOLD, PURPLE])
axes[2].set_title('Revenue Share by Category')

plt.suptitle('Sales Performance by Category', fontsize=13, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()

print('\nCategory Summary:')
print(cat_summary[['category', 'revenue', 'profit', 'margin_pct']].to_string(index=False))
print('\n📌 Margin % shows how efficiently each category converts revenue to profit.')
```

```python
# ─── CHART 3: Top 10 Products by Revenue & Gross Profit ────────────────────
prod_summary = (
    sales.groupby('product_name')
    .agg(revenue=('total_sale','sum'), profit=('gross_profit','sum'), units=('quantity_sold','sum'))
    .reset_index()
    .sort_values('revenue', ascending=False)
    .head(10)
)

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Top 10 by Revenue
axes[0].barh(prod_summary['product_name'], prod_summary['revenue'], color=BLUE, alpha=0.85)
axes[0].set_title('Top 10 Products by Revenue')
axes[0].set_xlabel('Total Revenue ($)')
axes[0].xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
axes[0].invert_yaxis()

# Top 10 by Gross Profit (sorted separately)
prod_profit = (
    sales.groupby('product_name')
    .agg(profit=('gross_profit','sum'))
    .reset_index()
    .sort_values('profit', ascending=False)
    .head(10)
)
axes[1].barh(prod_profit['product_name'], prod_profit['profit'], color=GREEN, alpha=0.85)
axes[1].set_title('Top 10 Products by Gross Profit')
axes[1].set_xlabel('Gross Profit ($)')
axes[1].xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
axes[1].invert_yaxis()

plt.suptitle('Product Performance Rankings', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

print('\n📌 Revenue rank ≠ profit rank. A high-revenue product with low margins may not deserve priority.')
```

```python
# ─── CHART 4: Physical vs Online Channel Performance ────────────────────────
channel = (
    sales.groupby('store_channel')
    .agg(
        revenue=('total_sale', 'sum'),
        profit=('gross_profit', 'sum'),
        orders=('sale_id', 'count'),
        avg_order=('total_sale', 'mean'),
    )
    .reset_index()
)
channel['margin_pct'] = (channel['profit'] / channel['revenue'] * 100).round(1)

labels = channel['store_channel'].tolist()
colors_ch = [RED, BLUE]

fig, axes = plt.subplots(1, 4, figsize=(16, 5))
metrics = [
    ('revenue',   'Total Revenue ($)',   mticker.FuncFormatter(lambda x,_: f'${x:,.0f}')),
    ('orders',    'Number of Orders',    None),
    ('avg_order', 'Avg Order Value ($)', mticker.FuncFormatter(lambda x,_: f'${x:.2f}')),
    ('margin_pct','Gross Margin (%)',    mticker.FuncFormatter(lambda x,_: f'{x:.1f}%')),
]
for ax, (col, ylabel, fmt) in zip(axes, metrics):
    ax.bar(labels, channel[col], color=colors_ch, width=0.5)
    ax.set_title(ylabel, fontsize=9)
    if fmt:
        ax.yaxis.set_major_formatter(fmt)

plt.suptitle('Physical vs Online Channel Comparison', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

print('\nChannel Breakdown:')
print(channel[['store_channel','revenue','orders','avg_order','margin_pct']].to_string(index=False))
print('\n📌 Even if one channel has fewer orders, a higher avg order value can make it more valuable per visit.')
```

```python
# ─── CHART 5: Customer Demographics ────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# Age distribution
axes[0].hist(customers['age'], bins=10, color=PURPLE, edgecolor='white', alpha=0.85)
axes[0].axvline(customers['age'].mean(), color=RED, linestyle='--',
                label=f'Mean age: {customers["age"].mean():.0f}')
axes[0].set_title('Customer Age Distribution')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Count')
axes[0].legend()

# Gender split
gender_counts = customers['gender'].value_counts()
gender_labels = ['Female' if g == 'F' else 'Male' for g in gender_counts.index]
axes[1].pie(gender_counts.values, labels=gender_labels, autopct='%1.1f%%',
            colors=[RED, BLUE], startangle=90)
axes[1].set_title('Gender Distribution')

# Customer segment breakdown
seg_counts = customers['segment'].value_counts()
axes[2].bar(seg_counts.index, seg_counts.values,
            color=[BLUE, GREEN, ORANGE, PURPLE, TEAL])
axes[2].set_title('Customers by Segment')
axes[2].set_ylabel('Number of Customers')
axes[2].tick_params(axis='x', rotation=20)

plt.suptitle('Customer Demographics & Segmentation', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

print('\nSegment Breakdown:')
print(seg_counts)
```

```python
# ─── CHART 6: Expense Breakdown ─────────────────────────────────────────────
exp_summary = expenses.groupby('category')['amount'].sum().sort_values(ascending=False)
exp_total   = exp_summary.sum()

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Pie chart
axes[0].pie(exp_summary.values, labels=exp_summary.index,
            autopct='%1.1f%%', startangle=90,
            colors=[BLUE, RED, GREEN, ORANGE, PURPLE, GOLD, TEAL, GREY])
axes[0].set_title('Expenses by Category (Share)')

# Bar chart with absolute amounts
axes[1].barh(exp_summary.index, exp_summary.values,
             color=[BLUE, RED, GREEN, ORANGE, PURPLE, GOLD, TEAL, GREY])
axes[1].set_title('Expenses by Category (Amount)')
axes[1].set_xlabel('Total Spent ($)')
axes[1].xaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'${x:,.0f}'))
axes[1].invert_yaxis()

plt.suptitle('Business Expense Analysis — 18 Months', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

print('\nExpense Totals:')
for cat, amt in exp_summary.items():
    print(f'  {cat:<30} ${amt:>12,.2f}  ({amt/exp_total*100:.1f}%)')
print(f'  {"TOTAL":<30} ${exp_total:>12,.2f}')
```

---
## STEP 4 — Diagnostic Analysis
### 🔍 Why did things happen the way they did?

Descriptive told us *what* happened. Diagnostic analysis helps us understand *why*.  
We look for patterns, correlations, and root causes behind the numbers.


```python
# ─── CHART 7: Seasonality — Monthly Revenue by Category ────────────────────
# Pivot: rows = month, columns = category
seasonal = (
    sales.groupby(['year_month', 'category'])['total_sale']
    .sum()
    .reset_index()
    .pivot(index='year_month', columns='category', values='total_sale')
    .fillna(0)
)
seasonal.index = pd.to_datetime(seasonal.index)
seasonal = seasonal.sort_index()

fig, ax = plt.subplots(figsize=(14, 6))
for col, color in zip(seasonal.columns, [BLUE, RED, GREEN, GOLD, PURPLE]):
    ax.plot(seasonal.index, seasonal[col], marker='o', label=col,
            color=color, linewidth=2, markersize=4)

ax.set_title('Monthly Revenue by Category — Seasonality Patterns', fontsize=13, fontweight='bold')
ax.set_ylabel('Revenue ($)')
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'${x:,.0f}'))
ax.legend(title='Category')
ax.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()

print('\n📌 Winter categories (Jackets, Hoodies) should spike Oct-Feb.')
print('   Summer categories (Dresses, Swimwear) should peak Apr-Aug.')
print('   If this pattern does NOT appear, it may indicate stock or pricing issues.')
```

```python
# ─── CHART 8: Promotion Impact on Revenue and Profit ───────────────────────
promo_agg = (
    sales.groupby('is_promotion')
    .agg(
        avg_sale=('total_sale', 'mean'),
        avg_profit=('gross_profit', 'mean'),
        total_orders=('sale_id', 'count'),
        total_revenue=('total_sale', 'sum'),
        total_profit=('gross_profit', 'sum'),
    )
    .reset_index()
)
promo_agg['margin_pct'] = (promo_agg['total_profit'] / promo_agg['total_revenue'] * 100).round(1)
promo_agg['label'] = promo_agg['is_promotion'].map({True: 'Promotion', False: 'No Promotion'})

labels = promo_agg['label'].tolist()
colors_p = [GOLD if v else GREY for v in promo_agg['is_promotion']]

fig, axes = plt.subplots(1, 3, figsize=(14, 5))

axes[0].bar(labels, promo_agg['avg_sale'], color=colors_p, width=0.5)
axes[0].set_title('Avg Sale Value')
axes[0].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'${x:.2f}'))

axes[1].bar(labels, promo_agg['avg_profit'], color=colors_p, width=0.5)
axes[1].set_title('Avg Gross Profit per Order')
axes[1].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'${x:.2f}'))

axes[2].bar(labels, promo_agg['margin_pct'], color=colors_p, width=0.5)
axes[2].set_title('Gross Margin %')
axes[2].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'{x:.1f}%'))

plt.suptitle('Impact of Promotions on Sales & Profitability', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

print('\nPromotion Summary:')
print(promo_agg[['label','avg_sale','avg_profit','total_orders','margin_pct']].to_string(index=False))
print('\n📌 Key question: Do promotions bring in ENOUGH extra orders to offset the margin reduction?')
```

## STEP 5 — Predictive Analysis
### 🔮 What is likely to happen next?

Predictive analysis uses patterns in historical data to forecast future outcomes.

We build **three models**, each answering a different business question:

| Model | Question | Technique |
|-------|----------|-----------|
| **Model 1: Revenue Forecast** | How much will we earn in the next 3 months? | Seasonal Linear Regression (Fourier) |
| **Model 2: Return Prediction** | Which transactions are likely to be returned? | Random Forest Classifier |
| **Model 3: Customer LTV** | How much is each customer worth over their lifetime? | Random Forest Regressor |

> **Why these models?**  
> Linear regression alone scored R²=0.04 — it completely missed the seasonal pattern.  
> Adding Fourier terms (sine/cosine waves) to capture seasonality improved R² to **0.63**.  
> Random Forest was chosen for return prediction because the dataset is imbalanced (~5% returns)  
> and tree-based models handle mixed feature types without scaling. LTV uses a regression forest  
> because spend is continuous and driven by a small number of strong predictors (order count, avg order value).

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import LeaveOneOut, cross_val_score
from sklearn.metrics import (
    r2_score, mean_absolute_error, mean_absolute_percentage_error,
    classification_report, confusion_matrix, ConfusionMatrixDisplay
)
import matplotlib.ticker as mticker

print('✅ Modelling libraries imported')
```

---
### 📈 MODEL 1 — Monthly Revenue Forecast
#### Seasonal Linear Regression with Fourier Terms

**Why not plain linear regression?**  
Revenue is driven by seasonality — jackets sell in winter, dresses in summer.  
A straight trend line ignores this and gives a near-zero R² (0.04) on our data.

**Fourier terms** let us encode the seasonal wave mathematically:
- `sin(2π × month/12)` and `cos(2π × month/12)` capture the **annual cycle**
- Adding a second harmonic (`sin(4π × month/12)`) captures **mid-year swings**

This gives us a trend line that **bends with the seasons** rather than ignoring them.

**Evaluation method:** Leave-One-Out Cross-Validation (LOO-CV)  
LOO-CV is best for small datasets (18 months). It trains on 17 months, tests on 1, rotates through all months — giving an honest estimate of out-of-sample error.


```python
# ─── MODEL 1: Seasonal Revenue Forecast ────────────────────────────────────

# Step 1: Aggregate to monthly revenue
monthly = (
    sales.groupby('year_month')['total_sale']
    .sum()
    .reset_index()
)
monthly['period']    = pd.to_datetime(monthly['year_month'])
monthly              = monthly.sort_values('period').reset_index(drop=True)
monthly['month_idx'] = range(1, len(monthly) + 1)          # sequential time index
monthly['month_num'] = monthly['period'].dt.month           # 1–12 for seasonality

# Step 2: Create Fourier features to capture seasonal patterns
# A full 12-month wave: sin and cos at frequency 1 and frequency 2
monthly['sin1'] = np.sin(2 * np.pi * monthly['month_num'] / 12)
monthly['cos1'] = np.cos(2 * np.pi * monthly['month_num'] / 12)
monthly['sin2'] = np.sin(4 * np.pi * monthly['month_num'] / 12)
monthly['cos2'] = np.cos(4 * np.pi * monthly['month_num'] / 12)

feature_cols = ['month_idx', 'sin1', 'cos1', 'sin2', 'cos2']
X = monthly[feature_cols]
y = monthly['total_sale']

# Step 3: Fit the model
model_rev = LinearRegression()
model_rev.fit(X, y)
y_fitted = model_rev.predict(X)

# Step 4: Evaluate — in-sample metrics
r2_in  = r2_score(y, y_fitted)
mae_in = mean_absolute_error(y, y_fitted)
mape_in = mean_absolute_percentage_error(y, y_fitted) * 100

# Step 5: LOO Cross-Validation for honest out-of-sample error
loo = LeaveOneOut()
loo_mae_scores = -cross_val_score(
    LinearRegression(), X, y,
    cv=loo, scoring='neg_mean_absolute_error'
)
loo_mae = loo_mae_scores.mean()

print('=' * 52)
print('  MODEL 1: Seasonal Revenue Forecast — Results')
print('=' * 52)
print(f'  In-sample R²:              {r2_in:.3f}')
print(f'  In-sample MAE:             ${mae_in:,.0f}/month')
print(f'  In-sample MAPE:            {mape_in:.1f}%')
print(f'  LOO Cross-Val MAE:         ${loo_mae:,.0f}/month   ← generalisation estimate')
print('=' * 52)

if r2_in >= 0.6:
    print('✅ Good seasonal fit. LOO-MAE is the realistic forecast error range.')
elif r2_in >= 0.4:
    print('⚠️  Moderate fit. Predictions are directional — real error ≈ LOO-MAE.')
else:
    print('❌ Weak fit. More data or a more complex model would help.')

print(f'\n📌 Interpretation: On average, our monthly forecast is off by ±${loo_mae:,.0f}.')
print(f'   With avg monthly revenue of ~${y.mean():,.0f}, that is ±{loo_mae/y.mean()*100:.1f}% error.')
```

```python
# ─── CHART 14: Revenue Forecast — Actual vs Fitted + 3-Month Forecast ──────

# Generate future months (Jul, Aug, Sep 2025)
last_idx  = monthly['month_idx'].max()
future_months = []
for i in range(1, 4):
    future_month_num = ((monthly['month_num'].iloc[-1] - 1 + i) % 12) + 1
    future_months.append({
        'month_idx': last_idx + i,
        'month_num': future_month_num,
        'sin1': np.sin(2 * np.pi * future_month_num / 12),
        'cos1': np.cos(2 * np.pi * future_month_num / 12),
        'sin2': np.sin(4 * np.pi * future_month_num / 12),
        'cos2': np.cos(4 * np.pi * future_month_num / 12),
    })
future_df  = pd.DataFrame(future_months)
future_rev = model_rev.predict(future_df[feature_cols])
future_labels = ['Jul 2025', 'Aug 2025', 'Sep 2025']

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# ── Left: Historical fit ──────────────────────────────────────────────────
ax = axes[0]
ax.bar(monthly['year_month'], y, color=BLUE, alpha=0.55, label='Actual Revenue', width=0.7)
ax.plot(monthly['year_month'], y_fitted, color=RED, linewidth=2.5,
        marker='o', markersize=5, label='Seasonal Trend (fitted)')

# Error bands (±LOO-MAE)
ax.fill_between(monthly['year_month'],
                y_fitted - loo_mae, y_fitted + loo_mae,
                alpha=0.15, color=RED, label=f'±${loo_mae:,.0f} error band (LOO-MAE)')

ax.set_title('Seasonal Model — Historical Fit', fontsize=12, fontweight='bold')
ax.set_ylabel('Monthly Revenue ($)')
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'${x:,.0f}'))
ax.tick_params(axis='x', rotation=45)
ax.legend(fontsize=9)

# ── Right: Forecast ───────────────────────────────────────────────────────
ax2 = axes[1]
all_labels = list(monthly['year_month']) + future_labels
all_actuals = list(y) + [None, None, None]
all_fitted  = list(y_fitted) + list(future_rev)

ax2.bar(monthly['year_month'], y, color=BLUE, alpha=0.55, label='Actual Revenue', width=0.7)
ax2.bar(future_labels, future_rev, color=GOLD, alpha=0.85, label='3-Month Forecast', width=0.7)
ax2.plot(all_labels, all_fitted, color=RED, linewidth=2, linestyle='--', marker='o',
         markersize=4, label='Model Trend')

# Uncertainty on forecast bars
for i, (lbl, rev) in enumerate(zip(future_labels, future_rev)):
    ax2.errorbar(lbl, rev, yerr=loo_mae, fmt='none', color='black',
                 capsize=6, capthick=2, linewidth=2)

ax2.axvline(x=len(monthly['year_month']) - 0.5, color='grey',
            linestyle=':', linewidth=2, label='Forecast boundary')
ax2.set_title('3-Month Revenue Forecast', fontsize=12, fontweight='bold')
ax2.set_ylabel('Monthly Revenue ($)')
ax2.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'${x:,.0f}'))
ax2.tick_params(axis='x', rotation=45)
ax2.legend(fontsize=9)

plt.suptitle('MODEL 1: Seasonal Revenue Forecast (Fourier Linear Regression)',
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

print('\n3-Month Forecast:')
for lbl, rev in zip(future_labels, future_rev):
    print(f'  {lbl}: ${rev:,.0f}  (±${loo_mae:,.0f})')
print('\n📌 Gold bars = forecast. Error bars show the realistic uncertainty range from LOO-CV.')
print('   These predictions assume seasonal patterns repeat — actual results will vary with promotions.')
```

### 🔄 MODEL 2 — Return Prediction
#### Random Forest Classifier

**Business question:** Can we predict which transactions are likely to be returned *before* the return happens?  
If yes, we can proactively reach out, improve product descriptions, or flag high-risk orders for review.

**Why Random Forest?**
- The dataset is **imbalanced** — only ~5% of transactions are returns
- Random Forest handles imbalance well with `class_weight='balanced'`
- It works with mixed numeric and encoded categorical features without scaling
- It gives us **feature importance** — telling us *what drives* returns

**Evaluation metrics:**
- **Precision** — of the returns we *predicted*, how many were actually returned?
- **Recall** — of all *actual* returns, how many did we catch?
- **F1 Score** — the balance between precision and recall (most important for imbalanced data)
- Accuracy alone is misleading here — predicting "never return" would give 95% accuracy but be useless


```python
# ─── MODEL 2: Return Prediction (Random Forest Classifier) ─────────────────

# Step 1: Build feature set
# We can only use features available AT TIME OF SALE (no post-sale info)
df_cls = sales[['unit_price', 'discount_pct', 'quantity_sold',
                 'month_num', 'category', 'store_channel', 'is_returned']].dropna().copy()

df_cls['day_of_week'] = sales['date'].dt.weekday
df_cls['is_weekend']  = (df_cls['day_of_week'] >= 5).astype(int)

# Encode categoricals
le_cat = LabelEncoder()
le_ch  = LabelEncoder()
df_cls['category_enc']    = le_cat.fit_transform(df_cls['category'])
df_cls['store_channel_enc'] = le_ch.fit_transform(df_cls['store_channel'])

feature_names = ['unit_price', 'discount_pct', 'quantity_sold',
                 'month_num', 'day_of_week', 'is_weekend',
                 'category_enc', 'store_channel_enc']

X_cls = df_cls[feature_names]
y_cls = df_cls['is_returned'].astype(int)

print(f'Dataset: {len(X_cls):,} transactions  |  Returns: {y_cls.sum():,} ({y_cls.mean()*100:.1f}%)')
print(f'Class imbalance ratio: 1 return per {int(1/y_cls.mean())} transactions\n')

# Step 2: Train Random Forest with class weighting to handle imbalance
model_ret = RandomForestClassifier(
    n_estimators=200,
    max_depth=6,
    class_weight='balanced',   # upweights the minority class (returns)
    random_state=42
)
model_ret.fit(X_cls, y_cls)
y_pred_cls = model_ret.predict(X_cls)
y_prob_cls = model_ret.predict_proba(X_cls)[:, 1]   # probability of return

# Step 3: Evaluate
print('=== Classification Report ===')
print(classification_report(y_cls, y_pred_cls, target_names=['No Return', 'Return']))

# Feature importance
fi = pd.Series(model_ret.feature_importances_, index=feature_names).sort_values(ascending=False)
print('\nFeature Importance (what drives return probability):')
for feat, imp in fi.items():
    bar = '█' * int(imp * 50)
    print(f'  {feat:<25} {bar}  {imp:.3f}')
```

```python
# ─── CHART 15: Return Model — Confusion Matrix & Feature Importance ─────────

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# ── Left: Confusion Matrix ────────────────────────────────────────────────
cm = confusion_matrix(y_cls, y_pred_cls)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['No Return', 'Return'])
disp.plot(ax=axes[0], colorbar=False, cmap='Blues')
axes[0].set_title('Confusion Matrix\nReturn Prediction Model', fontsize=12, fontweight='bold')

# Annotate with rate
tn, fp, fn, tp = cm.ravel()
axes[0].set_xlabel(
    f'True Negatives: {tn:,}  |  False Positives: {fp:,}\n'
    f'False Negatives: {fn:,}  |  True Positives: {tp:,}',
    fontsize=9
)

# ── Right: Feature Importance ─────────────────────────────────────────────
fi_sorted = fi.sort_values()
colors_fi = [GREEN if f in ['discount_pct','unit_price'] else BLUE for f in fi_sorted.index]
axes[1].barh(fi_sorted.index, fi_sorted.values, color=colors_fi)
axes[1].set_title('Feature Importance\nWhat Drives Return Probability?', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Importance Score')
axes[1].axvline(1/len(fi), color='grey', linestyle='--', label='Equal importance baseline')
axes[1].legend(fontsize=9)

plt.suptitle('MODEL 2: Return Prediction (Random Forest Classifier)',
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

print('\n📌 Reading the confusion matrix:')
print(f'   We correctly identified {tp:,} of {tp+fn:,} actual returns (recall = {tp/(tp+fn)*100:.0f}%)')
print(f'   We incorrectly flagged {fp:,} non-returns as returns (false alarm rate = {fp/(fp+tn)*100:.1f}%)')
print('\n📌 Feature importance insight:')
print(f'   "{fi.index[0]}" is the strongest predictor of returns.')
print('   Higher-value items and deeper discounts → higher return risk.')
print('   Practical use: Flag high-risk orders for extra quality checking before dispatch.')
```

### 💰 MODEL 3 — Customer Lifetime Value (LTV) Prediction
#### Random Forest Regressor

**Business question:** Given what we know about a customer today, how much are they likely to spend with us over their lifetime?

**Why this matters:**  
Not all customers are equal. Knowing predicted LTV lets us:
- Prioritise retention efforts on high-LTV customers
- Set appropriate acquisition cost limits (don't spend $100 acquiring a $50 customer)
- Personalise offers by predicted value tier

**Why Random Forest (not Linear Regression)?**  
LTV is driven by a small number of highly dominant features (`avg_order_value`, `num_orders`).  
Linear regression handles this fine, but Random Forest captures **non-linear interactions**  
(e.g., a Loyal customer with high avg order value is worth *far* more than the sum of those two factors individually).

**Features used — all observable at a customer profile level:**
- `age`, `gender`, `segment`, `num_orders`, `avg_order_value`, `tenure_days`


```python
# ─── MODEL 3: Customer Lifetime Value Prediction (Random Forest Regressor) ──

customers_clean = customers.copy()
customers_clean['first_purchase_date'] = pd.to_datetime(customers_clean['first_purchase_date'])
customers_clean['last_purchase_date']  = pd.to_datetime(customers_clean['last_purchase_date'])

# Only model customers with at least 1 order — new customers with 0 orders have no LTV signal
cust_active = customers_clean[customers_clean['num_orders'] > 0].copy()

# Step 1: Feature engineering
cust_active['tenure_days'] = (
    cust_active['last_purchase_date'] - cust_active['first_purchase_date']
).dt.days.fillna(0)
cust_active['gender_enc']  = (cust_active['gender'] == 'F').astype(int)
cust_active['segment_enc'] = LabelEncoder().fit_transform(cust_active['segment'])

feature_names_ltv = ['age', 'gender_enc', 'segment_enc',
                     'num_orders', 'avg_order_value', 'tenure_days']

X_ltv = cust_active[feature_names_ltv].fillna(0)
y_ltv = cust_active['total_spent']

print(f'Customers used for LTV modelling: {len(cust_active)}')
print(f'LTV range: ${y_ltv.min():,.0f} – ${y_ltv.max():,.0f}')
print(f'Avg LTV: ${y_ltv.mean():,.0f}\n')

# Step 2: Train model
model_ltv = RandomForestRegressor(
    n_estimators=200,
    max_depth=8,
    min_samples_leaf=3,
    random_state=42
)
model_ltv.fit(X_ltv, y_ltv)
y_ltv_pred = model_ltv.predict(X_ltv)

# Step 3: Evaluate
r2_ltv   = r2_score(y_ltv, y_ltv_pred)
mae_ltv  = mean_absolute_error(y_ltv, y_ltv_pred)
mape_ltv = mean_absolute_percentage_error(y_ltv, y_ltv_pred) * 100

# LOO-CV for honest error (smaller sample — manageable)
loo_ltv_mae = -cross_val_score(
    RandomForestRegressor(n_estimators=100, random_state=42),
    X_ltv, y_ltv, cv=LeaveOneOut(), scoring='neg_mean_absolute_error'
).mean()

print('=== LTV Regression Results ===')
print(f'  In-sample R²:         {r2_ltv:.3f}')
print(f'  In-sample MAE:        ${mae_ltv:,.2f}')
print(f'  In-sample MAPE:       {mape_ltv:.1f}%')
print(f'  LOO CV MAE:           ${loo_ltv_mae:,.2f}  ← realistic out-of-sample error')

fi_ltv = pd.Series(model_ltv.feature_importances_, index=feature_names_ltv).sort_values(ascending=False)
print('\nFeature Importance:')
for feat, imp in fi_ltv.items():
    bar = '█' * int(imp * 50)
    print(f'  {feat:<20} {bar}  {imp:.3f}')

# Add predictions back to customer table
cust_active['predicted_ltv'] = y_ltv_pred
cust_active['ltv_tier'] = pd.qcut(y_ltv_pred, q=3, labels=['Low', 'Mid', 'High'])
```

```python
# ─── CHART 16: LTV Model — Actual vs Predicted & Feature Importance ─────────

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# ── Left: Actual vs Predicted scatter ─────────────────────────────────────
colors_tier = {'High': GOLD, 'Mid': BLUE, 'Low': GREY}
for tier, grp in cust_active.groupby('ltv_tier'):
    axes[0].scatter(grp['total_spent'], grp['predicted_ltv'],
                    color=colors_tier[tier], alpha=0.6, s=40, label=f'{tier} LTV tier')

# Perfect prediction line
max_val = max(y_ltv.max(), y_ltv_pred.max())
axes[0].plot([0, max_val], [0, max_val], color=RED, linestyle='--',
             linewidth=1.5, label='Perfect prediction')
axes[0].set_title('Actual vs Predicted LTV', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Actual LTV ($)')
axes[0].set_ylabel('Predicted LTV ($)')
axes[0].xaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'${x:,.0f}'))
axes[0].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'${x:,.0f}'))
axes[0].legend(fontsize=9)
axes[0].set_title(f'Actual vs Predicted LTV
(R²={r2_ltv:.3f}, LOO MAE=±${loo_ltv_mae:,.0f})',
                  fontsize=11, fontweight='bold')

# ── Middle: Feature Importance ─────────────────────────────────────────────
fi_ltv_sorted = fi_ltv.sort_values()
axes[1].barh(fi_ltv_sorted.index, fi_ltv_sorted.values,
             color=[GOLD if v > 0.3 else BLUE if v > 0.1 else GREY for v in fi_ltv_sorted.values])
axes[1].set_title('LTV Feature Importance
What predicts customer value?',
                  fontsize=11, fontweight='bold')
axes[1].set_xlabel('Importance Score')
axes[1].axvline(1/len(fi_ltv), color='grey', linestyle='--', label='Equal baseline')
axes[1].legend(fontsize=9)

# ── Right: LTV distribution by tier ────────────────────────────────────────
for tier, color in colors_tier.items():
    subset = cust_active[cust_active['ltv_tier'] == tier]['predicted_ltv']
    axes[2].hist(subset, bins=15, color=color, alpha=0.6, label=f'{tier} LTV', edgecolor='white')
axes[2].set_title('Predicted LTV Distribution by Tier', fontsize=11, fontweight='bold')
axes[2].set_xlabel('Predicted LTV ($)')
axes[2].set_ylabel('Number of Customers')
axes[2].xaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'${x:,.0f}'))
axes[2].legend(fontsize=9)

plt.suptitle('MODEL 3: Customer Lifetime Value Prediction (Random Forest Regressor)',
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

# Summary by tier
tier_summary = cust_active.groupby('ltv_tier').agg(
    count=('customer_id','count'),
    avg_predicted_ltv=('predicted_ltv','mean'),
    avg_actual_ltv=('total_spent','mean'),
).round(2)
print('\nLTV Tier Summary:')
print(tier_summary.to_string())

print('\n📌 Dots close to the red diagonal = accurate predictions.')
print('   avg_order_value and num_orders dominate — these are the two levers to pull.')
print('   Increasing repeat visits (loyalty programme) or basket size (upsell) both directly grow LTV.')
```

### 📝 PREDICTIVE ANALYSIS — KEY FINDINGS

| Model | Metric | Result | Interpretation |
|-------|--------|--------|----------------|
| **Revenue Forecast** | R² (in-sample) | 0.63 | Good seasonal fit — far better than plain linear (0.04) |
| **Revenue Forecast** | LOO-CV MAE | ~$4,800/month | Realistic forecast error — use as uncertainty band |
| **Return Prediction** | Recall (returns) | ~98% | Model catches nearly all actual returns |
| **Return Prediction** | Precision (returns) | ~70% | Some false alarms — acceptable for a flagging system |
| **LTV Prediction** | R² | 0.996 | Excellent — driven by `avg_order_value` and `num_orders` |
| **LTV Prediction** | LOO-CV MAE | ~$44 | Very reliable for tier classification |

**Model Limitations to acknowledge in a portfolio:**
- Revenue forecast uses linear seasonality — a **SARIMA or Prophet model** would better handle irregular spikes (Black Friday, etc.) with more data
- Return prediction is **trained and tested on the same data** — for production, a proper train/test split over time is needed
- LTV model has very high R² because `avg_order_value × num_orders ≈ total_spent` by definition — the model is largely learning a formula. For a true LTV prediction, you'd predict *future* spend from *current* observable features only

> These are honest, appropriate caveats to include — they show analytical maturity to recruiters.

## STEP 6 — Prescriptive Analysis
### 💡 What should we DO about it?

Prescriptive analysis is the most actionable — it turns all findings into **concrete recommendations**.  
Instead of just saying "sales are declining," we say: *"Here's exactly what to do to fix it."*


```python
# ─── ACTION 1: Product Priority Scoring ─────────────────────────────────────
# Score each product on revenue contribution AND gross margin
# This helps decide: where should we invest stock, marketing, and shelf space?

prod_perf = (
    sales.groupby('product_name')
    .agg(
        total_revenue=('total_sale', 'sum'),
        total_profit=('gross_profit', 'sum'),
        total_units=('quantity_sold', 'sum'),
        num_orders=('sale_id', 'count'),
    )
    .reset_index()
)
prod_perf['margin_pct'] = prod_perf['total_profit'] / prod_perf['total_revenue']

# Normalize to 0-1
def normalize(series):
    return (series - series.min()) / (series.max() - series.min())

prod_perf['rev_score']    = normalize(prod_perf['total_revenue'])
prod_perf['margin_score'] = normalize(prod_perf['margin_pct'])
prod_perf['vol_score']    = normalize(prod_perf['total_units'])

# Weighted composite: 50% revenue, 30% margin, 20% volume
prod_perf['priority_score'] = (
    0.50 * prod_perf['rev_score'] +
    0.30 * prod_perf['margin_score'] +
    0.20 * prod_perf['vol_score']
).round(3)

prod_perf = prod_perf.sort_values('priority_score', ascending=False).reset_index(drop=True)

fig, ax = plt.subplots(figsize=(11, 8))
colors_score = [GREEN if s >= 0.6 else ORANGE if s >= 0.35 else RED
                for s in prod_perf['priority_score']]
ax.barh(prod_perf['product_name'], prod_perf['priority_score'], color=colors_score)
ax.set_title('Product Priority Score\n(Green ≥ 0.6 = Star | Orange = Manage | Red = Review)', fontsize=12, fontweight='bold')
ax.set_xlabel('Priority Score (0–1)')
ax.invert_yaxis()
plt.tight_layout()
plt.show()

stars = prod_perf[prod_perf['priority_score'] >= 0.6]['product_name'].tolist()
low   = prod_perf[prod_perf['priority_score'] <  0.35]['product_name'].tolist()
print(f'⭐ STAR products (invest more): {stars}')
print(f'⚠️  Low priority (review/reduce): {low}')
```

```python
# ─── ACTION 2: Customer Segmentation & VIP Identification ───────────────────
# Who are the most valuable customers? Identify and retain them.

cust_sales = (
    sales[sales['customer_id'] != 'GUEST']
    .groupby('customer_id')
    .agg(
        total_spent=('total_sale', 'sum'),
        num_orders=('sale_id', 'count'),
        last_date=('date', 'max'),
    )
    .reset_index()
)

# RFM-style: rank by total_spent
cust_sales = cust_sales.sort_values('total_spent', ascending=False).reset_index(drop=True)

# Top 20% = VIP
vip_cutoff = int(len(cust_sales) * 0.20)
cust_sales['tier'] = 'Regular'
cust_sales.loc[:vip_cutoff-1, 'tier'] = 'VIP'

vip_rev = cust_sales[cust_sales['tier'] == 'VIP']['total_spent'].sum()
reg_rev = cust_sales[cust_sales['tier'] == 'Regular']['total_spent'].sum()
total_cust_rev = vip_rev + reg_rev

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Revenue share pie
axes[0].pie([vip_rev, reg_rev], labels=['VIP (Top 20%)', 'Regular (Bottom 80%)'],
            autopct='%1.1f%%', colors=[GOLD, BLUE], startangle=90, explode=(0.05, 0))
axes[0].set_title('Revenue: VIP vs Regular Customers')

# Scatter: orders vs spend, coloured by tier
colors_tier = [GOLD if t == 'VIP' else BLUE for t in cust_sales['tier']]
axes[1].scatter(cust_sales['num_orders'], cust_sales['total_spent'],
                c=colors_tier, alpha=0.7, s=50)
axes[1].set_title('Customer Spend vs Number of Orders')
axes[1].set_xlabel('Number of Orders')
axes[1].set_ylabel('Total Spent ($)')
axes[1].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'${x:,.0f}'))

from matplotlib.lines import Line2D
legend_els = [Line2D([0],[0],marker='o',color='w',markerfacecolor=GOLD,markersize=10,label='VIP'),
              Line2D([0],[0],marker='o',color='w',markerfacecolor=BLUE,markersize=10,label='Regular')]
axes[1].legend(handles=legend_els)

plt.suptitle('Customer Value Segmentation', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

print(f'Total customers analysed:  {len(cust_sales):,}')
print(f'VIP customers (top 20%):   {vip_cutoff:,}')
print(f'VIP revenue:               ${vip_rev:,.2f}  ({vip_rev/total_cust_rev*100:.1f}% of total)')
print(f'Avg VIP spend:             ${cust_sales[cust_sales["tier"]=="VIP"]["total_spent"].mean():,.2f}')
print(f'Avg Regular spend:         ${cust_sales[cust_sales["tier"]=="Regular"]["total_spent"].mean():,.2f}')
print('\n📌 Launch a VIP loyalty programme (early access, exclusive discounts, free shipping)')
print('   to retain your highest-value customers — they generate disproportionate revenue.')
```

```python
# ─── ACTION 3: Optimal Discount Strategy ────────────────────────────────────
# What is the "sweet spot" discount that drives volume WITHOUT destroying margin?

disc_roi = (
    sales.groupby('discount_tier', observed=True)
    .agg(
        total_revenue=('total_sale', 'sum'),
        total_profit=('gross_profit', 'sum'),
        order_count=('sale_id', 'count'),
    )
    .reset_index()
)
disc_roi['margin_pct']   = (disc_roi['total_profit'] / disc_roi['total_revenue'] * 100).round(1)
disc_roi['avg_order_val'] = (disc_roi['total_revenue'] / disc_roi['order_count']).round(2)
disc_roi['roi_score']    = (disc_roi['margin_pct'] * disc_roi['order_count'] / disc_roi['order_count'].max()).round(1)

colors_d2 = [GREEN, GOLD, ORANGE, RED]
d_labels  = disc_roi['discount_tier'].astype(str).tolist()

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].bar(d_labels, disc_roi['margin_pct'], color=colors_d2)
avg_margin = disc_roi['margin_pct'].mean()
axes[0].axhline(avg_margin, color='black', linestyle='--', label=f'Avg: {avg_margin:.1f}%')
axes[0].set_title('Gross Margin % by Discount Tier')
axes[0].set_ylabel('Margin (%)')
axes[0].legend()
axes[0].tick_params(axis='x', rotation=15)

axes[1].bar(d_labels, disc_roi['order_count'], color=colors_d2)
axes[1].set_title('Orders per Discount Tier')
axes[1].set_ylabel('Number of Orders')
axes[1].tick_params(axis='x', rotation=15)

axes[2].bar(d_labels, disc_roi['avg_order_val'], color=colors_d2)
axes[2].set_title('Avg Order Value per Discount Tier')
axes[2].set_ylabel('Avg Order ($)')
axes[2].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'${x:.2f}'))
axes[2].tick_params(axis='x', rotation=15)

plt.suptitle('Discount Strategy Optimisation', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

print('\nDiscount Tier Summary:')
print(disc_roi[['discount_tier','order_count','avg_order_val','margin_pct']].to_string(index=False))
print('\n📌 Recommendation: Cap promotions at the tier that keeps margin % above the average.')
print('   Never run 20% blanket discounts — they destroy margin without proportional order uplift.')
```

```python
# ─── ACTION 4: Inventory Reorder Dashboard ──────────────────────────────────
# Which products need restocking NOW?

latest_stock = (
    inventory.sort_values('date')
    .groupby('product_id')
    .last()
    .reset_index()
)[['product_id', 'product_name', 'stock_level', 'reorder_point', 'supplier', 'stock_value']]

latest_stock['status'] = latest_stock.apply(
    lambda r: '🔴 REORDER NOW' if r['stock_level'] <= r['reorder_point']
              else ('🟡 LOW STOCK' if r['stock_level'] <= r['reorder_point'] * 1.5
              else '🟢 OK'),
    axis=1
)

reorder_now = latest_stock[latest_stock['status'] == '🔴 REORDER NOW']

fig, ax = plt.subplots(figsize=(13, 7))
colors_stock = [RED if s <= rp else ORANGE if s <= rp*1.5 else GREEN
                for s, rp in zip(latest_stock['stock_level'], latest_stock['reorder_point'])]
bars = ax.barh(latest_stock['product_name'], latest_stock['stock_level'], color=colors_stock)
ax.axvline(latest_stock['reorder_point'].mean(), color='orange', linestyle='--',
           linewidth=2, label=f'Reorder Point ({int(latest_stock["reorder_point"].mean())} units)')
ax.set_title('Current Stock Levels — Live Reorder Dashboard', fontsize=13, fontweight='bold')
ax.set_xlabel('Units in Stock')
ax.legend()
ax.invert_yaxis()
plt.tight_layout()
plt.show()

print(f'\n⚠️  Products requiring immediate reorder: {len(reorder_now)}')
print(reorder_now[['product_name','stock_level','reorder_point','supplier','status']].to_string(index=False))
```

```python
# ─── ACTION 5: Best Day & Channel for Promotions ────────────────────────────
# Where and when do customers spend the most? Target promotions precisely.

day_channel = (
    sales.groupby(['day_name', 'store_channel'])
    .agg(avg_sale=('total_sale', 'mean'), orders=('sale_id', 'count'))
    .reset_index()
)
day_channel.columns = ['day', 'channel', 'avg_sale', 'orders']

day_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
online   = day_channel[day_channel['channel']=='Online'].set_index('day')['avg_sale'].reindex(day_order).fillna(0)
physical = day_channel[day_channel['channel']=='Physical'].set_index('day')['avg_sale'].reindex(day_order).fillna(0)

x = np.arange(len(day_order))

fig, ax = plt.subplots(figsize=(13, 5))
ax.bar(x - 0.2, physical.values, width=0.4, color=RED, label='Physical', alpha=0.85)
ax.bar(x + 0.2, online.values,   width=0.4, color=BLUE, label='Online', alpha=0.85)
ax.set_xticks(x)
ax.set_xticklabels(day_order, rotation=15)
ax.set_title('Avg Revenue per Transaction by Day & Channel', fontsize=13, fontweight='bold')
ax.set_ylabel('Avg Revenue per Sale ($)')
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'${x:.2f}'))
ax.legend(title='Channel')
plt.tight_layout()
plt.show()

# Best combinations
best_combo = day_channel.sort_values('avg_sale', ascending=False).head(3)
print('\n🎯 Top 3 Day + Channel combinations (highest avg sale):')
print(best_combo[['day','channel','avg_sale','orders']].to_string(index=False))
print('\n📌 Run flash sales on these specific day+channel combos.')
print('   Customers are already in a buying mindset — a small promotion goes a long way.')
```

```python
# ─── ACTION 6: Segment-Level Marketing Budget Allocation ────────────────────
# How much does each customer segment contribute? Where should the marketing $ go?

seg_perf = (
    sales[sales['customer_id'] != 'GUEST']
    .merge(customers[['customer_id','segment']], on='customer_id', how='left')
    .groupby('segment')
    .agg(
        revenue=('total_sale', 'sum'),
        profit=('gross_profit', 'sum'),
        orders=('sale_id', 'count'),
    )
    .reset_index()
)
seg_perf['avg_order'] = (seg_perf['revenue'] / seg_perf['orders']).round(2)
seg_perf['margin_pct']= (seg_perf['profit'] / seg_perf['revenue'] * 100).round(1)
seg_perf = seg_perf.sort_values('revenue', ascending=False)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

seg_colors = [GOLD, BLUE, GREEN, ORANGE, PURPLE]

axes[0].bar(seg_perf['segment'], seg_perf['revenue'], color=seg_colors)
axes[0].set_title('Revenue by Customer Segment')
axes[0].set_ylabel('Revenue ($)')
axes[0].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'${x:,.0f}'))
axes[0].tick_params(axis='x', rotation=15)

axes[1].bar(seg_perf['segment'], seg_perf['avg_order'], color=seg_colors)
axes[1].set_title('Avg Order Value by Segment')
axes[1].set_ylabel('Avg Order ($)')
axes[1].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'${x:.2f}'))
axes[1].tick_params(axis='x', rotation=15)

plt.suptitle('Customer Segment Performance', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

print('\nSegment Performance:')
print(seg_perf[['segment','revenue','avg_order','margin_pct','orders']].to_string(index=False))
print('\n📌 Budget allocation insight:')
print('   - Loyal: Reward with exclusive perks (highest LTV, protect at all costs)')
print('   - Occasional: Nudge with personalised re-engagement emails')
print('   - New: Welcome offers to convert to Loyal (first 90 days are critical)')
print('   - Seasonal: Time campaigns to their active windows (summer/winter)')
print('   - Bargain Hunter: Offer early access to sales — avoid costly full-price ads')
```
