## Q1: What's the Most Profitable Product?
**Analytics type: Descriptive**

We measure profitability at two levels:
- **Net profit per transaction** (`net_profit`) ~ actual realised profit after returns.
- **Profit margin %** from the products catalogue ~ to see which products earn more per dollar sold.

```python
# Total net profit and units sold per product
product_profit = sales.groupby('product_name').agg(
    total_net_profit = ('net_profit', 'sum'),
    total_units_sold = ('quantity_sold', 'sum'),
    total_revenue    = ('total_sale', 'sum'),
    transactions     = ('sale_id', 'count')
).reset_index()

product_profit['avg_profit_per_unit'] = (product_profit['total_net_profit'] / product_profit['total_units_sold']).round(2)
product_profit = product_profit.sort_values('total_net_profit', ascending=False)

print("Top 10 Most Profitable Products (by total net profit):")
print(product_profit.head(10)[['product_name','total_net_profit','total_units_sold','avg_profit_per_unit']].to_string(index=False))
```
